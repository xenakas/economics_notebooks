## 5. Валютные кризисы

- Защита фиксированного курса может принести больше вреда, чем пользы 
- Переключение валютного режима с привязки (любого типа) к плавающему курсу 
- Резкая номинальная девальвация внутренней валюты к иностранной
- Резкое улучшение баланса текущего счета
- Большой отток капитала и потеря валютных резервов ЦБ

Много других часто встречающихся, но не всегда
обязательных фактов – падение ВВП, длительный
спад кредитования. Следствие резкого изменения
условий на валютных рынках

Доверие к режиму и уровень резервов:
- Технически невозможно сохранять ФР без резервов
- Следовательно, низкие уровни резервов приведут к низкому доверию рынка
- Увеличивая резервы по сравнению с денежной базой, ЦБ может повысить доверие к ФР, и дать инвесторам уверенность, что режим останется на долгое время
- Обещания держать курс без подкрепления наличием резервов не имеют веса (имеет значение «доступ» ЦБ к резервам , а не фактическая сумма)

### Кризис доверия к ФР в модели Дорнбуша

Кризис доверия - инвесторы и д/хозяйства ожидают будущую девальвацию

- $S^e$ растет,  т.е. ожидания рынка будут отличаться от обещаний правительства
- но правительство не признает возможность будущей девальвации и утверждает, что обменный курс будет навсегда равен $S_0$
- Когда $S_e$ увеличивается, UIP  сдвигается вверх, заставляя ЦБ продавать резервы и уменьшать денежную массу для поддержки курса
- Степень недоверия можно понять по наблюдаемой разнице местной и иностранной процентных ставок

### Устойчивость ФР


1. Уровень резервов
2. Устойчивость банковской системы



- Неблагоприятный сдвиг ожиданий в экономике при ФР будет растить внутренние процентные ставки из-за сокращения предложения денег 
- Когда эти события достаточно радикальны, то есть отток капитала + потеря ЗВР + рост процентных ставок достаточно велики, их принято называть «спекулятивная атака против национальной валюты» 


###   Модель валютного кризиса

####  Модель 1-го поколения  – плохая политика правительства

- Ожидания девальвации растут, потому что рынок понимает, что правительство ведет политику, несовместимую с ФР 
- Модели могут объяснить, почему спекулятивная атака происходил лавинообразно

Примеры:

1. Рецессия (то есть Y падает), и ЦБ проводит денежную экспансию в надежде, что она (каким-то чудом) будет эффективной, а девальвации не произойдет
2. Гос-во берет деньги у ЦБ для финансирования дефицита бюджета
3. Рынок ожидает, что денежная масса должна расти в будущем из-за I/II


####  Модель 2-го поколения – самореализующиеся ожидания

- В этих моделях если бы не спекулянты, то правительство могло бы и дальше проводить политику, согласующуюся с ФР 
- Но спекулятивные ожидания девальвации делают защиту ФР столь дорогостоящей в плане трат ЗВР, что правительство сдается и отказывается от дальнейшей защиты

#### Модели 3-го поколения – банковская система

- Защита ФР могла бы продолжаться и дальше, резервов достаточно
- Но из-за защиты ставки повышаются до неприемлемого для банков уровня, и правительство вынуждено отказаться от защиты
