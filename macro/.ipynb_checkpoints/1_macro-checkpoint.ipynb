### Введение

Кейнсианцы (теория совокупного спроса): 
- Совокупный спрос определяет равновесное состояние в экономике
- Вмешательство государства необходимо
- Стабилизационная политика государства:  антициклическая политика спроса

Неоклассики (теория совокупного предложения):
- Все рынки являются совершенно конкурентными. Все цены абсолютно гибкие. Вывели макроэкономическое поведение через агрегацию микроэкономического поведения агентов.
- Совокупное предложение определяет равновесное состояние в экономике, поскольку политики спроса неэффективны.
- Раскрыли причины неэффективного воздействия политик спроса в условиях абсолютно гибких цен. 
- Для стабилизации экономики государство не должно вмешиваться в ее функционирование.


Монетаризм (M. Friedman, A. Schwartz):
- Объяснение output-inflation tradeoff в краткосрочном периоде и ее отсутствие в долгосрочной перспективе(существование SRAS и LRAS)
    - Гипотеза естественного уровня безработицы
    - Моделирование инфляционных ожиданий: адаптивные инфляционные ожидания 
- Реинтерпретация количественной теории
    - Правило постоянного темпа роста денежной массы
- Все рынки являются совершенно конкурентными, все цены абсолютно гибкие.
- Рыночному механизму приспособления мешает неправильные инфляционные ожидания экономических агентов в краткосрочном периоде, обладающие системной ошибкой (адаптивные инфляционные ожидания)
- Политики спроса эффективны в краткосрочном периоде и неэффективны в долгосрочном.
- Монетарная политика  эффективней фискальной политики , так при проведении последней присутствует значительный эффект вытеснения.
- Стабилизация экономики достигается через стабилизацию самой политики

Новый классический подход (R. Lucas, T. Sargent, N. Wallace, R. Hall):
- Объяснение output-inflation tradeoff в краткосрочном периоде и ее отсутствие в долгосрочной перспективе(существование SRAS и LRAS) при отсутствие системной ошибки
- Неэффективность ожидаемой макроэкономической политики
    - Любые систематические изменения в макроэкономической политике не воздействуют на реальную экономику
    - Только непредвиденные изменения в политике могут воздействовать на реальную экономику в краткосрочной перспективе
- «Критика Лукаса»: 
    - на экономику существенное влияние оказывают  ожидания экономических агентов, формируемые на основе будущего. 
    - важные изменения в политике государства, воздействующие на будущее, могут нарушить зависимость, наблюдаемую в прошлом, или даже поменять ее на противоположную
- Все рынки являются совершенно конкурентными, все цены абсолютно гибкие.
- Рыночному механизму приспособления мешает неправильные инфляционные ожидания экономических агентов в краткосрочном периоде, обладающие случайной ошибкой (рациональные  инфляционные ожидания)
- Политики спроса эффективны в краткосрочном периоде и неэффективны в долгосрочном.
- Стабилизация экономики достигается через неожиданную политику спроса.



Модель AD-AS позволяет: 
1. выявить условия макроэкономического равновесия, определить 
    величину равновесного объема производства и равновесного уровня цен; 
2. объяснить колебания объема производства и уровня цен в экономике; показать причины и последствия этих изменений; 
3. описать различные варианты экономической политики государства 

### Кейнсианский крест

Предпосылки: 
- выпуск  ниже своего потенциального уровня (Y < Y*);
- уровень цен постоянный (P = const);
- совокупное предложение совершенно эластично (кривая AS горизонтальна) ‒ фирмы могут произвести столько товаров, сколько они знают покупатели захотят купить);
- заработная плата жесткая («липкая» по терминологии Кейнса; если безработица высока, рабочие не могут требовать повышения заработной платы (W = const), поэтому цены ресурсов не меняются;
- ставка процента постоянна (R = const);
- национальный выпуск = национальному доходу (это означает, что амортизация и косвенные налоги не учитываются);
- налоги только прямые и платят их только домохозяйства;
- экономика закрытая (NX = 0) (однако методология анализа позволяет использовать эту модель и для условий открытой экономики).


Объяснение отрицательного наклона кривой совокупного спроса: 3 эффекта. Посколь-
ку совокупный спрос есть сумма всех спросов в экономике, зависящих от дохода, то его отри-
цательный наклон нельзя объяснить микроэкономическим обоснованием. 

Есть три эффекта:

1. Эффект Пигу (эффект реального богатства). При повышении уровня цен реальное богатство индивидов уменьшается, в результате чего люди чувствуют себя беднее и меньше
потребляют. Поскольку потребительский спрос есть крупная составляющая совокупного
спроса, совокупный спрос отрицательно зависит от уровня цен.
2. Эффект процентной ставки (эффект Кейнса). С ростом цен людям нужно больше наличных денег, чтобы приобретать товары и услуги, поэтому дополнительные деньги снимаются с банковских счетов. При неизменном денежном предложении банкам сложнее
выдавать кредиты, соответственно кредиты становятся дороже, то есть растет процентная ставка, или растет стоимость денег. Фирмы меньше инвестируют, поскольку деньги
становятся дороже, а так как инвестиции являются частью совокупного спроса, он в свою
очередь снижается, то есть агрегированный спрос отрицательно зависит от уровня цен.
3. Эффект чистого экспорта. С увеличением цен для иностранцев местные цены становятся
более высокими, что ведет к падению экспорта. С другой стороны для сограждан импортные цены становится ниже, что ведет к росту импорта. Так, чистый экспорт снижается и
как следствие, совокупный спрос, частью которого является чистый экспорт.

Вывод с точки зрения Кейнса: воздействовать на экономику надо (преимущественно
через гос. закупки), самая бесполезная политика — политика сбалансированного бюджета (речь
про фискальную политику).

Виды фискальной политики

По характеру воздействия на доход:
- Стимулирующая – увеличивающая доход
    - проводится в период спада;
    - применяется с целью    стимулирования экономической деятельности;
    - нацелена на увеличение ВВП  и снижение безработицы; 
    - предполагает: увеличение  госзакупок, увеличение трансфертов, снижение налогов,  равное увеличение G и Tx (стимулирующий сбалансированный бюджет).

- Сдерживающая – уменьшающая доход
    - проводится в период бума,  когда экономика перегрета;
    - применяется с целью   уменьшения реального ВВП;
    - нацелена на снижение уровня цен и борьбу с инфляцией; 
    - предполагает: снижение госзакупок, уменьшение трансфертов, повышение налогов,   равное уменьшение G и Tx   (сдерживающий сбалансированный бюджет).

По способу проведения:
- Дискреционная – проводимая субъективно, по решению кого-либо
    - предполагает официально объявляемое изменение государственных расходов и/или налогах, которые являются  результатом изменений в политике правительства, с целью изменить уровень совокупных расходов и преодолеть рецессионный или инфляционный разрыв
- Автоматическая – самопроизвольно проводимая через встроенные в экономику автоматические стабилизаторы политика (налог на прибыль,  прогрессивный подоходный налог, пособия по безработице)
    - основана на действии в экономике автоматических (встроенных стабилизаторов) – инструментов, которые автоматически уменьшают экономические колебания и помогают стабилизировать выпуск, автоматически изменяя чистые налоговые поступления, когда экономика движется по направлению от или к) уровню выпуска полной занятости.

### Финансовый рынок и монетарная политика

Финансовые активы
- Денежные  активы (M)
    - Спрос на деньги = предложению денег
- Неденежные финансовые активы: акции (E) и облигации (B)
    - Спрос на ценные бумаги = предложению ценных бумаг
    
    
Финансовый рынок в равновесии, если и денежный рынок и  рынок неденежных активов в равновесии.

Закон Вальраса для финансового рынка:
- Равновесие одного из сегментов финансового рынка (денежного рынка) означает равновесие другого сегмента (рынка облигаций) и равновесие всего финансового рынка в целом, что облегчает анализ и позволяет ограничиться изучением равновесия только на денежном рынке.
Если на денежном рынке неравновесие, то восстановление равновесия происходит за счет  изменения ситуации на рынке облигаций.

Норма обязательных резервов
- Основа системы частичного  банковского резервирования
- Определяет способность банковской системы создавать депозиты
- Устанавливается  центральным банком 

Создание денег коммерческими банками

Необходимым условием для создания денег коммерческими  банками  является  наличие  системы  частичного  резервирования.

Условия максимального увеличения  предложения денег банками :
- Банки имеют только обязательные резервы
- Денежное обращение происходит только в безналичной форме

Банковский (депозитный) мультипликатор показывает величину предложения денег, созданную 1 денежной единицей депозитов.

Спрос на деньги 

Существует три мотива для хранения денег:
- трансакционный мотив – деньги необходимы для совершения сделок, т.е. покупки товаров и услуг;
- мотив предосторожности (предусмотрительный) – деньги нужны для незапланированных (непредвиденных) покупок, т.е. обусловлен неопределенностью;
- спекулятивный мотив – деньги – это финансовый актив, но  существуют другие виды финансовых активов (акции и облигации), которые могут приносят процентный доход, изменение ставки процента меняет не только спрос на акции и облигации, но и на деньги.

Теория предпочтения ликвидности

Предложена Дж.М.Кейнсом, который:
- добавил к трансакционному мотиву спроса на деньги спекулятивный мотив и мотив предосторожности, обосновав обратную зависимость величины спроса на деньги от ставки процента;
- рассматривал спрос на деньги как спрос на «реальные денежные балансы» mD,  который определяется:  реальным доходом/выпуском Y и  номинальной ставкой процента  R (альтернативными        издержками хранения денег), изменение которой ведет к изменению структуры финансового портфеля;
- исходил из предпосылки жестких цен (Р = сonst), т.е. отсутствия инфляции, поэтому номинальная ставка процента R  совпадает с реальной ставкой процента r (R= r)

Монетарная политика – политика проводимая ЦБ посредством изменения денежного предложения.

Инструменты монетарной политики
- изменение нормы обязательных резервов 
    - Воздействует на кредитные возможности банков
    - Воздействует на денежный и банковский мультипликатор
    - Если норма резервирования растет, кредитные возможности первого банка падают и банковский мультипликатор уменьшается, денежное предложение уменьшается, на денежном рынке возникает избыток спроса, на рынке ценных бумаг дефицит спроса, ставка процента растет, инвестиции падают, расходы падают, доходы падают.
- изменение учетной ставки процента (ставки рефинансирования) 
    - Воздействует на кредитные возможности банков
    - Денежный и банковский мультипликаторы (при прочих равных условиях) остаются неизменными
    - Если ставка рефинансирования растет, кредитные возможности банков падают и, при том же банковском мультипликаторе, денежное предложение уменьшается, на денежном рынке возникает избыток спроса, на рынке ценных бумаг дефицит спроса, ставка процента растет, инвестиции падают, расходы падают, доходы падают.
- операции на открытом рынке 
    - Государство финансирует увеличение дефицита бюджета за счет займа у населения, денежная база и денежная масса при этом остаются неизменными (фискальная политика)
    - При покупке (продажи) ЦБ гособлигаций  активы ЦБ растут (падают), пассивы ЦБ растут(падают), денежная база и денежная масса растут (падают). (Монетарная политика)


Рост денежного предложения ведет к дефициту спроса на денежном рынке и избытку спроса на рынке ценных бумаг при том же уровне дохода, ставка процента падает, спрос на деньги растет и денежный рынок приходит в равновесие при более низкой ставке процента , но при том же уровне дохода. То есть линия   LM сдвигается вниз. 

### IS-LM

Модель IS-LM: общая характеристика

- Модель была предложена английским экономистом Джоном Хиксом в 1937 году, дополнена в 1949 году Элвина Хансена . 

- Модель IS-LM является теоретической базой для объяснения краткосрочных колебаний экономики, т.е. делового цикла. 

Модель IS-LM 
- представляет собой модель совместного равновесия  товарного  и денежного рынков;
- является моделью кейнсианского типа;
- описывает поведение экономики в краткосрочном периоде; 
- позволяет проанализировать воздействие на экономику фискальной и монетарной политики и сравнить их эффективность;
- служит основой современной теории  совокупного спроса.

Модель IS-LM сохраняет все предпосылки простой Кейнсианской модели (модели Кейнсианского креста): 
- уровень цен  Р = cоnst;
- номинальная заработная плата W = cоnst;
- совокупное предложение AS совершенно эластично;
- условие равновесия товарного рынка: Y = AEпл  и  Iпл= Sнац. ;
- экзогенные переменные: государственные закупки G, денежная масса M, аккордные налоги Tx, трансферты Tr, налоговая ставка t
- эндогенные переменные: совокупный выпуск Y, потребление C, инвестиции I.

Отличие: ставка процента R ≠ const, т.е. R  становится эндогенной переменной и определяется изменением ситуации (равновесия) на денежном рынке. 


Кривая IS (Investment = Saving) описывает равновесие товарного рынка и показывает все возможные комбинации уровней ставки процента R и реального совокупного дохода (выпуска) Y, при которых товарный рынок находится в равновесии, т.е. когда доход (выпуск) равен планируемым расходам  (Y = AЕпл.), инъекции равны изъятиям, а планируемые инвестиции в закрытой экономике  равны нацинальным сбережениям  (Iпл. = Sнац.). 


Кривая LM (Liquidity preference = Money supply) описывает равновесие денежного рынка и показывает все возможные комбинации уровней ставки процента R и реального дохода Y, при которых денежный рынок  находится в равновесии, т.е. реальный спрос на деньги (предпочтение ликвидности) равен реальному предложению денег (М/Р)D = (МS/Р).
- Рост выпуска Y ведет к росту трансакционного и поэтому общего спроса на реальные денежные балансы mD, что при фиксированном реальном предложении денег ведет к повышению ставки процента R Таким образом, более высокому выпуску соответствует более высокая ставка процента. 

Неравновесие товарного рынка и модель IS

- Если  экономика находится не на кривой IS, то товарный рынок находятся в ситуации неравновесия, которая ликвидируется экономикой изменением объема производства.
- Если экономика находится в точке лежащей над IS, то в экономике товарный рынок находится в состоянии  дефицита спроса, что заставляет фирмы сокращать производство, пока оно не станет равно спросу.
- Если экономика находится в точке лежащей под IS, то в экономике товарный рынок находится в состоянии  избытка спроса, что заставляет фирмы наращивать производство, пока оно не станет равно спросу.


Неравновесие денежного рынка и модель LM

- Если  экономика находится не на кривой LM, то денежный рынок и рынок ценных бумаг находятся в ситуации неравновесия, которая ликвидируется экономикой изменением ставки процента.
- Если экономика находится в точке лежащей над LM, то в экономике денежный рынок находится в состоянии  дефицита спроса, а рынок ценных бумаг в состоянии избытка спроса, что заставляет ставку процента падать, восстанавливая равновесие и на рынке денег и на рынке ценных бумаг.
- Если экономика находится в точке лежащей под LM, то в экономике денежный рынок находится в состоянии  избытка спроса, а рынок ценных бумаг в состоянии дефицита спроса, что заставляет ставку процента расти, восстанавливая равновесие и на рынке денег и на рынке ценных бумаг.

Равновесие товарного рынка за счет изменения фирмами запасов непроданной продукции и поэтому выпуска Y. Равновесие денежного рынка за счет изменения ставки процента R. Быстрее равновесие устанавливается на денежном рынке (движение из т.А в т.В), так как для этого достаточно изменить структуру портфеля финансовых активов, что не требует значительных затрат времени. Часто предполагается мгновенное установление равновесия на денежном рынке (движение вдоль кривой LM  из точки В в точку Е).


Изменение равновесия и переход к новому общему равновесию в модели IS-LM происходит, если меняется:
- величина автономных расходов АЕ0 (параллельный сдвиг кривой IS);
- мультипликатор автономных расходов и/или чувствительность инвестиций к ставке процента (изменение наклона кривой IS);
- величина реального предложения денег (параллельный сдвиг кривой    LM);
- чувствительность спроса на деньги к доходу и/или чувствительность спроса на деньги к ставке процента (изменением наклона кривой LM). 


- Эффект вытеснения  в модели IS-LM это разница между изменением дохода в кейнсианском кресте и изменением дохода в модели IS-LM в результате проведения стимулирующей фискальной политики.
- Уменьшение дохода в модели IS-LM  в традиционной кейнсианской модели происходит из-за снижения инвестиционных расходов в результате роста ставки процента, т.к. в этой модели только они зависят от ставки процента.
- Чем меньше эффект вытеснения тем больше изменение дохода в модели IS-LM в результате воздействия фискальной политики.


### Открытая экономика

- Валютный рынок обслуживает международные сделки по обмену товарами и услугами и движению капитала
- Спрос на валютном рынке на национальную валюту предъявляют нерезиденты, предложение – резиденты страны.
- Номинальный валютный курс в условиях плавающего валютного курса формирует валютный рынок, в условиях фиксированного валютного курса – ЦБ.

Валютный рынок (плавающий валютный курс)
- Рынок является совершенно конкурентным, равновесное состояние – это ситуация когда спрос на валюту равен предложению валюты. 
- В состояние равновесия устанавливается такой валютный курс, при котором сальдо платежного балансе равно 0.
- Равновесие восстанавливается рынком: в состоянии избытка (дефицита) спроса на рынке национальной валюты  для восстановления равновесия, национальная валюта начинает дорожать  (дешеветь) величина спроса на нац. валюту падает(растет).
- Изменение состояния равновесия на валютном рынке происходит при изменении спроса на валюту и (или) изменения предложения валюты.

Валютный рынок (фиксированный валютный курс)
- Рынок не является совершенно конкурентным, равновесие устанавливается ЦБ на объявленном им уровне валютного курса.
- Для восстановления равновесия ЦБ проводит валютные интервенции: в состоянии избытка (дефицита) спроса на рынке национальной валюты  ЦБ  увеличивает(продает) официальные резервы .
- Изменение состояния равновесия на валютном рынке происходит по решению ЦБ о девальвации (уменьшении фиксированного курса) или о  ревальвации (увеличения фиксированного курса).
- Валютные интервенции - вмешательство ЦБ в функционирование валютного рынка для  установления нужного ЦБ валютного курса, через изменение официальных резервов (OR)


Нестерилизованные интервенции ЦБ - покупая (продавая ) OR ЦБ меняет не только предложение отечественной валюты на валютном рынке, а, следовательно,  и валютный курс,  но и денежную базу , что ведет к изменению предложения денег в отечественной экономике.

Стерилизованные интервенции  ЦБ - ЦБ одновременно с покупкой (продажей ) OR  продает (покупает )гособлигации , что приводит к изменению предложения отечественной валюты на валютном рынке, а, следовательно,  и валютного курса, но не меняет денежную базу , то есть изменения предложения денег в отечественной экономике не происходит.


Спекулятивная атака - проводится при фиксированном валютном курсе при заниженной стоимости иностранной валюты (завышенной стоимости отечественной валюты)
- Цель атаки: заставить ЦБ отказаться от режима фиксированного валютного курса. В этом случае валютный рынок установит более высокий валютный курс иностранной валюты, по которому спекулянты смогут продать валюту , купленную по фиксированному заниженному курсу.
- Схема атаки: спекулянты увеличивают спрос на иностранную валюту, национальная валюта дешевеет, чтобы вернуть ее стоимость на фиксированный уровень ЦБ продает OR, предложение отечественной валюты падает, стоимость отечественной валюты возвращается на фиксированный уровень, спекулянты увеличивают спрос на иностранную валюту, национальная валюта дешевеет…….ЦБ отказывается от фиксации


Моделирование поведения валютного курса
- На основе равенства потоков: валютный курс уравновешивает потоки доходов, входящие в национальную экономику ( input),  и потоки доходов, уходящие из  национальной экономики (output). 
- На основе соотношения цен в мировой экономике: на основе абсолютного или относительного паритета покупательной способности.
- На основе соотношения доходностей национальных и иностранных активов: на основе непокрытого процентного паритета.



### IS-LM-BP

Основные предпосылки:
- валютный курс определяется равновесием спроса и предложения на валютном рынке; 
- изменения валютного курса происходит свободно; 
- центральный банк не проводит интервенций на валютном рынке. 

Основная идея: валютный курс плавающий и устанавливается таким образом, чтобы сальдо платежного баланса, который состоит из счета текущих операций (СА) и счета движения капитала (CF), было равно нулю:

ВР = 0  если  СА + CF = 0

Равновесный валютный курс определяется параметрами, от которых зависят СА и CF


- Рынок является совершенно конкурентным, равновесное состояние – это ситуация когда спрос на валюту равен предложению валюты. 
- В состояние равновесия устанавливается такой валютный курс, при котором сальдо платежного балансе равно 0.
- Равновесие восстанавливается рынком: в состоянии избытка (дефицита) спроса на рынке национальной валюты  для восстановления равновесия, национальная валюта начинает дорожать  (дешеветь) величина спроса на нац. валюту падает(растет).
- Изменение состояния равновесия на валютном рынке происходит при изменении спроса на валюту и (или) изменения предложения валюты.

Условие Маршала-Лернера: при прочих равных условиях реальное обесценение валюты вызывает улучшение баланса счета текущих операций, если объемы экспорта и импорта достаточно эластичны к изменению реального валютного курса.
- Удорожание нац. валюты влияет на стоимость экспорта  через два эффекта: эффект стоимости (ЭС) и эффект объема (ЭО). ЭС увеличивает стоимость чистого экспорта, так как стоимость одной единицы экспорта растет, ЭО уменьшает стоимость экспорта, так как объем экспортируемых товаров падает. Чтобы при удорожании нац.валюты стоимость чистого экспорта падала, надо чтобы  ЭО > ЭС



### Рынок труда

Предположения неоклассической модели
- Предложение определяет спрос 
- Процессы, происходящие в экономике, являются результатом решений миллионов индивидуумов

Фундаментом неоклассического анализа являются теоретические основы микроэкономики, с помощью которых изучается 
- формирование индивидуальных решений домашних хозяйств и фирм, 
- процесс функционирования домохозяйств и фирм, их взаимодействие в рыночной экономической системе. 

Методология анализа
- Описывается поведение репрезентативного экономического агента 
- Посредством агрегирования происходит переход к изучению общеэкономических  явлений и тенденций 


Предпосылки модели с гибкими ценами

- Все цены в экономике гибкие
- В экономике функционируют три макроэкономических рынка
- Все рынки являются рынками совершенной конкуренции
- Каждый макроэкономический агент состоит из множества экономических агентов, действующих одинаковым образом

 Особенности неоклассического рынка труда

- Труд является универсальным трудом, который может быть использован для производства любого товара. 
- Спрос на труд предъявляют фирмы, максимизируя прибыль.
- Предложение труда предъявляют домохозяйства, максимизируя полезность.
- Цена труда – реальная заработная плата определяется рынком труда на основе уравновешивания спроса и предложения  на труд.


Анализ спроса на труд
- Прибыль каждой фирмы описывается  через неоклассическую производственную функцию, задающую ее производство. 
- Фирма выбирает количество необходимого ей труда, решая задачу максимизации прибыли.
- Условию первого порядка задачи максимизации прибыли (предельный продукт труда равен реальной зарплате) определяет количество труда, необходимого фирме.
- Из условия первого порядка следует, что количество необходимого фирме труда отрицательно зависит от реальной заработной платы.
- Так спрос на труд всех фирм отрицательно зависит от реальной заработной платы, то и спрос на труд всей экономики отрицательно зависит от реальной заработной платы.


Теория жизненного цикла

Разработана Франко Модильяни ,Альбертом Андо и Ричардом Брумберг (1954 г.) 

К предпосылкам теории межвременного выбора добавляются следующие предпосылки:
- совокупный доход человек получает неравномерно: совокупный доход увеличивается, когда человек работает,  затем совокупный доход уменьшается,
- сбережения определяются желанием индивидуумов обеспечить себе   желаемый уровень потребления в старости, сбережения позволяют   перераспределять доход с периода, когда он высок, на период, когда он  низок или его вообще нет, 
- на протяжении всей жизни человек хочет поддержать единый уровень потребления.


Модель постоянного дохода

Модель разработана Милтоном Фридманом в 1957 г.
Основные предпосылки те же, что и в теории межвременного выбора, но кроме этого предполагается, что
- динамика дохода не носит такой предсказуемый характер, как в модели жизненного цикла, 
- потребители стараются поддержать единый уровень потребления в течении жизни,
- в текущем доходе можно выделить постоянный и временный доход.

Постоянный доход - доход, на который можно рассчитывать и в будущем.
Временный доход – доход, носящий временный, случайный характер.

### AD-AS

Кривая совокупного спроса – это кривая, каждая точка которой показывает сочетания величины совокупного спроса (совокупных планируемых расходов и поэтому дохода) Y и уровня цен Р, при которых денежный и товарный рынок, при прочих равных условиях, находятся в равновесии.   

Кривая AD может быть выведена из модели IS-LM, если «снять» предпосылку о постоянстве уровня цен. Если уровень цен Р падает (растет), то реальное предложение денег растет(падает), и кривая LM при том же выпуске сдвигается вправо(влево), в результате в модели IS-LM устанавливается новое равновесие, в котором равновесная ставка процента и равновесный доход растут (падают).
