## 3. Модели формирования номинального обменного курса


Отличия номинального курса (S)  от реального (RER): 
- Меньше влияние рынков товаров, больше – рынков финансовых активов (Колебания S влияют на номинальную доходность активов)
- Более короткий временной горизонт (В краткосрочном периоде цены жесткие, так что колебания S определяющие для движений RER)
- Связь с денежно-кредитной и курсовой политикой ЦБ

Долгосрочная перспектива (LR):
- Гибкие цены (меняются мгновенно) 
- Плавающий валютный курс
- Главное предположение: раз все гибкое, то выполняется ППС: $ S = \frac{P}{P^*}$

Равновесие на денежных рынках
$$ \frac{M_s}{P}=L(Y, i)$$

- $Y\uparrow \rightarrow L(Y, i)\uparrow $ рост спроса на деньги (кривая спроса на деньги целиком сдвигается вверх)
- $i\uparrow \rightarrow L(Y, i) \downarrow $ снижение спроса на деньги (движение вдоль кривой спроса на деньги)

Номинальный курс определяется относительным
денежным предложением и спросом в 2-х странах: 

$$ S = \frac{ P}{P^*} = \frac{M_s  L(Y^*, i^* )}{L(Y, i) M_s^* } $$

Предположим логарифмическую спецификацию функции спроса на деньги в 2-х странах, тогда: 

$$ m – p = \alpha  + \beta y – \gamma i \\ s = p – p^* = (m–m^* ) – \beta (y–y^* ) + \gamma (i–i^* ) $$ 


где $ m=\ln(M), y=\ln(Y) , p=\ln(P), s=\ln(S) $

Выводы:
- В странах, испытывающих существенный рост денежной массы  будет ослабление нац. валюты
$$ M_s \uparrow ( m\uparrow) \rightarrow P\uparrow \rightarrow S\uparrow \\ M_s^*\uparrow ( m^*\uparrow)\rightarrow P^*\uparrow \rightarrow S\downarrow $$
- Скорость роста цен зависит только от скорости роста предложения денег, а значит скорость изменения курса также зависит только от (относительной) скорости печатания денег
- В странах, испытывающих существенный экономический рост будет укрепление нац. валюты
- В странах, испытывающих существенный рост процентной ставки в LR  будет ослабление нац. валюты: 
$$ i \uparrow \rightarrow L(Y, i)\downarrow \rightarrow P\uparrow [ M_s =const, Y=const, i=\pi^e +r, r=const \Rightarrow  \pi^e \uparrow] \rightarrow S\uparrow \\ 
i^* \uparrow \rightarrow L(Y^*, i^*)\downarrow \rightarrow P^*\uparrow \rightarrow S\downarrow $$
- Влияние ставки i на валютный курс S будет отличаться в SR и LR, т.к.  в LR  изменение происходит  за счет ожидаемой инфляции $\pi^e$, в SR чаще за счет $M^s$

### Подход к курсу через рынок активов : UIP

Инвестиции в два актива – домашний и
иностранный: одинаковая срочность, одинаковая степень рискованности (дефолта), одинаковая ликвидность

$S_t$ - текущий курс рублей за 1 доллар

При условии свободного движения капитала и для
нейтральных к риску (risk-neutral) инвесторов:
- ожидаемая доходность от вложения в два актива, посчитанная с учетом изменения валютного курса, должна быть одинаковой
$$( 1+ i_t ) = (1 + i_t^* ) \frac{ S_{ t+1}^e}{S_t} $$

Если взять логарифмы от обеих частей ($\ln(1+x)\sim x$):
$$i_t = i_t^* + \left( \frac{ S_{ t+1}^e}{S_t}– 1 \right)$$ - это т.н. простая форма UIP  (Непокрытый паритет процентных ставок) 

$\frac{ S_{ t+1}^e}{S_t}– 1$  – ожидаемое ослабление курса рубля по
отношению к доллару

Если $i_t >i_t^* $, то ожидается, что домашняя
валюта будет дешеветь по отношению к иностранной
валюте

### Подход к курсу через рынок активов: CIP

CIP (покрытый паритет процентных ставок):

- Будущий рыночный курс $ S_t+1 $ неизвестен, и инвестор несет валютный риск 
- Чтобы его избежать, инвестор может подписать форвардный контракт по определенной цене $F_t$
- Тогда при отсутствии ограничений на движение капитала и при отсутствии арбитража выполняется равенство безрисковых доходностей двух активов:
$$(1+ i_t )= (1+ i_t^* )\frac{ F_t}{ S_t} \\ i_t = i_t^* + \left( \frac{ F_{ t}}{S_t}– 1 \right)$$

Если $ \frac{F_{ t}}{S_t}– 1 > 0 $, то говорят, что домашняя валюта
продается с форвардным дискаунтом (т.е. форвардный
курс домашней валюты слабее, чем спотовый курс) или
наоборот, иностранная валюта продается с форвардной
премией

CIP может не выполняться если:
- существуют ограничения на движение капитала (или угроза ввода)
- разные, по мнению рынка, риски дефолта национальных контрагентов

Премия на ограничение движения капитала или
страновая премия (CRP):
$$ (1+ i_t - СRP_t )= (1+ i_t^* ) \frac{F_t}{S_t} \\  CRP_t = i_t - i_t^* - \left(\frac{F_t}{S_t} -1\right) $$


- $ CRP_t >0$ ограничения на движение капитала не дают некоторым инвесторам зайти в домашние активы, доходность домашних активов могла бы стать ниже с приходом новых инвестров и притоком капитала
- $ CRP_t <0$ ограничения на движение капитала не дают держателям домашних активов легко «избавиться» от них, доходность домашних активов стала бы выше, если бы часть держателей смогла бы выйти 


Если UIP & CIP выполняются, то форвардный курс должен равняться ожидаемому курсу с учетом срочности:
$$F_{t,t+k} = E_t (S_{t+k} ) $$

Предположим, что форвардный курс на 1 год вперед доллара к евро равняется 1,56 долл. за евро, а текущий курс равен 1,3 долл. за евро. 

_Чему равна форвардная премия/форвардный дискаунт_

$$ FP =  F/S-1 = 1.56/1.3-1 = 0.2 $$ 

Обменные курсы квотируются в единицах долларов за одно евро, так что евро здесь - иностранная валюта, и торгуется с форвардной премией.

В простой форме: $ i - i^* = (F/S-1) = 0.2$ 




### Подход к курсу через рынок активов: RP

UIP должно быть модифицировано таким образом, чтобы учитывать валютный риск: чем более волатилен валютный курс, тем больше должна быть компенсация за риск 

- Для домашнего инвестора, покупающего иностранные бумаги, доходность на иностранные бумаги должна превосходить доходность на домашние бумаги
(рассчитанную в домашней валюте) на премию за риск $(RP_t >0)$:
$$ (1+ i_t )= (1+ i_t^* - RP_t ) \frac{S_{t+1}^e}{S_t} $$ 

- Для иностранного инвестора:  инвестирование в наши домашние бумаги должно превосходить доходность инвестирования в свои активы: 
$$ (1+ i_t -  RP_t^* )= (1+ i_t^*)\frac{S_{t+1}^e}{S_t} $$ 


Можем определить премию за риск $RP_t$ как отклонение от UIP: 
$$ RP_t = \left(\frac{S_{t+1}^e}{S_t} -1\right) - ( i_t - i_t^* ) $$

Если CIP выполняется (при отсутствии ограничений
на движение капитала), то
$$ i_t - i_t^* =\left(\frac{F_t}{S_t} -1\right) \\  RP_t = \left(\frac{S_{t+1}^e}{S_t} -1\right) - \left(\frac{F_t}{S_t} -1\right) = \frac{S_{t+1}^e - F_t}{S_t}$$
 

$RP_t$   - форвардное смещение (forward bias), отделяет ожидаемый курс валюты от форвардного курса. 

Ошибка прогноза спотового курса $\frac{S_{t+1} - S_{t+1}^e}{S_t} $ .

$$ \frac{S_{t+1} - F_{t}}{S_t} = \frac{S_{t+1}^e - F_t}{S_t} +   \frac{S_{t+1} - S_{t+1}^e}{S_t} $$ 

Форвардный курс может плохо предсказывать будущий спотовый курс из-за:
- существования премии за риск 
- ошибок рынка в прогнозе будущего спотового курса.



### Равновесие на валютном рынке в SR и LR

$$( 1+ i_t ) = (1 + i_t^* ) \frac{ S_{ t+1}^e}{S_t} \\ S_t = UIP(\cdot) = S_{t+1}^e \frac{1+ i_t^*}{1+ i_t} $$



Условие равновесия на валютном рынке (UIP): равенство ожидаемых доходностей двух активов,
посчитанных в одной валюте (UIP)



$$ i_t = UIP(\cdot)  = i_t^*  +  \frac{S_{t+1}^e}{S_{t+1} }-1  $$


где $i_t$ -  доходность домашнего актива



- $i_t \downarrow  \rightarrow  S_t \uparrow $: cнижаются национальные ставки $\Rightarrow$ национальный актив менее доходен $\Rightarrow$ инвесторы уходят в другие активы, то есть в другую валюту $\Rightarrow$ национальная валюта дешевеет




Валютный курс - вперед-смотрящая переменная: 
$$ S_t = \frac{1+i^*_t}{1+i_t} \dots \frac{1+i^{e*}_{t+n}}{1+i^{e}_{t+n}} E_t(S_{t+n+1}) $$

- $S_t$ реагирует на все новости и слухи о будущем
- Волатильная переменная, которую сложно предсказать
- На ежедневной основе можно представить $S_t$ как случайное блуждание (random walk): 
$$S_t = S_{t-1 } + e_t, где\ e_t \sim N(0, \sigma^2) \\  E_t [S_{t+k} ] = E_t [S_t + e_{t + 1} + \dots  + e_{t+k} ] = S_t $$







### Валютный курс через фундаментальные факторы: 

Номинальный курс = дисконтированная сумма ожиданий
функции от фундаментальных факторов

Из монетарного подхода:

$$ s_t = p_t – p_t^* = (m_t–m_t^* ) – \beta (y_t–y_t^* ) + \gamma (i_t–i^*_t ) $$ 


Из UIP:  
$$ i_t -  i_t^*  =   s_{ t+1}^e - s_t $$

Подставим в монетарный подход и получим: 

$$ s_t =\frac{(m_t–m_t^* ) – \beta (y_t–y_t^* ) + \gamma s_{ t+1}^e }{1 + \gamma} = Z_t + \alpha s_{ t+1}^e = Z_t + \alpha Z_{t+1}^e + ... + \alpha^{k-1} Z_{t+k-1}^e + \alpha^k  s_{ t+k}^e $$

$Z_t$  содержит в себе т.н. фундаментальные факторы (динамику денежной массы и экономический рост)

#####  Задача 

_Рассмотрим малую открытую экономику, в которой цены  - постоянно гибкие, а выпуск определяется экзогенно. Предположим, что спрос на деньги задан:_


$$ m_t – p_t =  y_t – \beta i_t $$


_ЦБ регулирует денежную массу как инструмент политики. Выполняются ППС и непокрытый паритет процентных ставок:_

$$ s_t = p_t – p^*_t \\  i_t - i^*_t = E_t(s_{t+1}-s_t)$$ 


_Покажите, что валютный курс может быть выражен как функция своего же ожидания в следующий момент времени  и экзогенных факторов, определяющих спрос на денежную массу и предложение денежной массы:_

$$  s_t = p_t – p^*_t =  m_t - y_t +  \beta i_t   – p^*_t =  m_t - y_t +  \beta (i_t - i_t^*) +  \beta i_t^*   – p^*_t  = m_t - y_t +  \beta E_t(s_{t+1}-s_t) +  \beta i_t^*   – p^*_t  $$

отсюда $$  s_t = \dfrac{m_t – p^*_t  - y_t +  \beta s_{t+1}^e  +  \beta i_t^*}{1+\beta}    $$


подставляем   $s_{t+1}^e$ из такой же формулы как сверху, но для периода t+1 и так далее, получаем искомую формулу бесконечной суммы: 

$$  s_t = \dfrac{1}{1+\beta} \sum \left( \dfrac{\beta}{1+\beta} \right)^{i^*} (m_{t+i}^e – p_{t+i}^{*e}  - y_{t+i}^e +   \beta i_{t+i}^e)$$



Чтобы формула была справедлива, сумма должна сходится: для этого необходимо и достаточно $\beta \geq 0$, то если множитель $ \dfrac{\beta}{1+\beta} < 1$ по модулю. 


_Как постоянное увеличение денежной массы влияет на изменение валютного курса в этой модели?_


$$ \Delta s = \dfrac{1}{1+\beta} \sum \left( \dfrac{\beta}{1+\beta} \right)^{i^*} \Delta m$$

То есть $ \Delta s  $  для всех моментов времени начиная с $t$ одинаково. Перелета нет, краткосрочный курс (близко к t) меняется также как и долгосрочный (далеко от t).


### Carry trade

Сarry trade - инвестиционная стратегия международных
инвесторов 
- занять в валюте с низкой $i_t $ 
- купить актив в валюте с высокой $i_t^* $ 


Если $s_{t+k}  - s_t = \alpha + \beta (i_{t,k} – i_{t,k}^* )+e_{t,t+k}$
и $\beta<1$,  то имеет
смысл инвестировать в валюту с более высокой
доходностью 

Особенно выгодна стратегия carry trade, когда
высокодоходная валюта продолжает укрепляться, при этом приток капитала в carry trade валюту
способствует ее укреплению

История изменений обменных курсов полна провалами кэрри-трейда. Утверждение о выгодности кэрри-трейда основано на эконометрических заключениях о среднем значении за длительные периоды.

Периоды, когда сarry trade приносит большую прибыль,
сменяется периодами больших убытков, когда
высокодоходная валюта начинает ослабевать или
начинает «лопаться пузырь», например, рынка
недвижимости – “unwinding of the carry trade”.




### Forward premium puzzle

Парадокс форвардной премии: $F_t$ – смещенная оценка $S_{t+1}$

Объяснения:
1. Не удается смоделировать столь волатильную премию за риск, какая наблюдается эмпирически
2. Участники рынка – не вполне рациональные агенты (noise trader)
3.  $S_{t+1}^e \neq  E(S_{t+1 })$,  например, если распределения шоков оценивается инвесторами отличным от экономистов образом
4. Peso-problem: сравнительно небольшая вероятность негативного сценария (резкого ослабления кэрри-трэйд валюты), которое не происходит в период доступной временной выборки
5. Simultaneity bias


### Денежный и валютный рынки в SR

На денежном рынке:  $ \frac{M^s}{P} = L(Y, i)$ 

На валютном рынке:   $( 1+ i_t ) = (1 + i_t^* ) \frac{ S_{ t+1}^e}{S_t} $ - доходность домашнено актива = ожидаемая доходность зарубежного актива

- $ M^s \uparrow $ (временно) $\rightarrow [M/P] \uparrow (P =const)  $ избыток предложения реального денежного баланса $\rightarrow i \downarrow (Y =const) \rightarrow  $ не выполняется UIP $ \rightarrow  $ снижение спроса на домашние и рост спроса на зарубежные активы, снижение спроса на рубль и рост спроса на доллар  $ \rightarrow   S\uparrow  [S_{t+1}^e =const]$  рубль слабеет к доллару, так что выполняется UIP: $ i= i^*+ [S_{t+1}^e / S -1] $

В долгосрочной перспективе (LR):
- $Y_{LR} = Y_p$ - устанавливается потенциальный (естественный) уровень выпуска,  естественный уровень безработицы
- Цены и зарплаты – гибкие, мгновенно меняются для поддержания равновесия на рынке товаров, услуг и труда




_При 10 % увеличении М: в долгосрочной перспективе выполняются ППС ( $S=P/P^* $) и соответствие роста цен росту денег ( $P=M/L(Y,i)$ ). Поэтому M, P и S все вырастут в одинаковой пропорции (10%). Краткосрочная жесткость цен тут совсем неважна._


### Модель «перелета» (overshooting) курса Дорнбуша


__Необратимое изменение денежной массы в LR:__
в SR валютный курс «перелетает» свое LR
значение за счет жесткости цен в SR

Является результатом жесткости цен в SR, выполнения
условия равновесия на валютном рынке (UIP) и
выполнения ППС в LR




Пусть $M^s \uparrow $ (permanent):

В SR: 

$[M/P]↑ $ (P=const) - избыток предложения реального
денежного баланса$ → i↓ →$спрос на активы$↓$, спроса на рубль$↓ →
S↑ $


Кроме того: 

$M^s ↑ $ (permanent) $→P^e_{LR} ↑ → $ (ППС в LR) $→ S_{LR}^e ↑ $ [сдвиг
кривой UIP] $→  S↑↑$ (больше, чем в случае временного $M^s ↑$, как
рассматривали до этого):

$$ S_{SR} = S_2 > S_3 - при \ временном \ M^s ↑$$


Пока реального сектора в модели нет
– предположим, что если $i↓ → Y↑, L(Y, i)↑$ [сдвиг кривой $L(Y, i)$] в SR, то сдвиг незначителен, т.е. можем пренебречь сдвигом кривой спроса на деньги: $i_{SR} < i_1$ - начальное значение


- LR:  $ M^s \uparrow $ (permanent) $P \uparrow [M/P]\downarrow (M =const) $ избыточный спроса на реальный денежный баланс $ \rightarrow i \uparrow \ rightarrow$ спрос на домашние активы $uparrow$, спрос на рубль $ \uparrow \rightarrow S\downarrow$, рубль укрепляется при переходе в LR
- $i$ возвращается к изначальному значению


В LR: 

$P^e_{LR} ↑ → P↑ →[M/P] ↓ (M =const) $ - избыточный спроса на
реальный денежный баланс $ → i↑ →$ спрос на домашние активы $↑$,
спрос на рубль$↑ → S↓$, рубль укрепляется при переходе в LR

$$ S_{LR} =S_4 < S_{SR} =S_2 $$ 

$$[S_{LR} - S_1 ] / S_1 = [P_{LR} - P_1 ] / P_1 = [M_{LR} - M_1 ] / M_1 $$ 

$i$ возвращается к изначальному значению

__Вывод: в SR валютный курс «перелетает» свое LR значение за счет жесткости цен в SR__

- Валютный курс демонстрирует «перелет», когда мгновенная (в SR) реакция валютного курса на шок больше, чем реакция в долгосрочной перспективе
- Эффект помогает объяснить краткосрочную волатильность валютного курса


### Случай «недолета» (undershooting) валютного курса

Влияние необратимого увеличения денежной массы в
домашней экономике (США): «недолет» валютного курса
(менее вероятен, чем «перелет»)


Пусть $M^s ↑$ (permanent)

В SR: 

$[M/P]↑ (P =const)$ - избыток предложения реального
денежного баланса$ → i↓ →$спрос на активы$↓$, спрос на рубль$↓$ →
$S↑$, рубль ослабевает

Кроме того: 

$M^s ↑ →P^e_{LR} ↑ → S_{LR}^e ↑ → S↑↑$ (больше, чем в случае
временного $M^s ↑$, как рассматривали до этого):
$$S_{SR} = S_2 > S_3 (при временном M^s ↑)$$

Если эффект на выпуск $i↓ → Y↑, L(Y, i)↑$ в SR значительный
– то происходит сдвиг кривой спроса на деньги: $i_{SR} > i_1$, т.е. эффект роста спроса на деньги (после увеличения
денежного предложения в SR) сильнее влияет на $i$, чем эффект
роста денежного предложения

В LR: 

$P ↑ →[M/P] ↓ (M =const) $ - избыточный спроса на реальный
денежный баланс в США $→ i↑ →$спрос на активы↑, спроса на $ \$ ↑$ →
$S↓$:
$$S_{LR} =S_4 > S_{SR} =S_2$$


$$[S_{LR} - S_1 ] / S_1 = [P_{LR} - P_1 ] / P_1 = [M_{LR} - M_1 ] / M_1 $$ 

i – возвращается к изначальному значению




__В SR валютный курс «недолетает» до своего LR значения
за счет более сильного влияния на $i$ эффекта роста спроса
на деньги, чем изначального эффекта $M^s ↑$ (+ жесткость
цен в SR)__


Влияния необратимого изменения темпа роста денежной
массы при гибких ценах

UIP в простой форме:
$$i_t - i_t^* = S_{t+1}^e / S_t -1$$

Продифференцируем ППС и возьмем ожидания в LR:
$$S_{t+1}^e / S_t -1 = \pi_t^e - \pi_t^{e*}$$

Эффект Фишера: 
$$i_t - i_t^* = \pi_t^e - \pi_t^{e*}$$


увеличение ожидаемого уровня
инфляции на $\Delta \pi^e$ ведет к увеличению $i$ также на $\Delta \pi^e$

##### Задача

_Предположим, что в открытой экономике модели Дорнбуша с жесткостью цен в краткосрочной перспективе произошло перманентное (необратимое)  снижение спроса на национальную валюту._

Подсказка: перманентное (необратимое) снижение спроса на национальную валюту - сдвиг кривой спроса на деньги. Влияние снижения спроса на нац. валюту на денежный рынок и цены (в долгосрочной перспективе) аналогично влиянию увеличению предложения нац. валюты (денежной массы). 

_Как при этом изменится валютный курс в краткосрочной и долгосрочной перспективе?_

Перелет- это когда краткосрочная реакция курса на шок оказывается больше (по модулю, но в том же направлении), чем долгосрочная. 

«Реакция» - это разница между курсом в новом равновесии и его начальным уровнем.

Начинаем решать с конца, чтобы понять, как изменятся ожидания. 

Меньше спроса на деньги навсегда означает, что в LR цены вырастут (т.к. $M/P=L(Y,i)$, а правая часть снижается при неизменном $М$). Рост цен означает, что домашняя валюта ослабнет номинально (т.к. $SP^* = P$ при неизменном $P^*$). То есть агенты будут ожидать ослабления курса в LR, соответственно $S^e$ вырастет.

На верхней части диаграммы кривая UIP уже в SR сдвинется вправо-вверх из-за роста параметра Se

На нижней части диаграммы в SR кривая md сдвинется влево-вверх из-за снижения спроса как функции. При этом предположения о «перелете» позволяют сделать вывод, что эффекты возможного изменения Y в SR будут малы, и не скажутся на направлении движения кривой $M^d$

Итого, в SR равновесии ставка задается пересечением нового $M^d$ и начального M/P, и проецируется на новую UIP. 

В LR равновесии цены вырастут так, чтобы ставка вернулась к начальной. 



#####   Задача

_Предположим, что в открытой экономике модели Дорнбуша с жесткостью цен в краткосрочной перспективе произошло перманентное (необратимое)  снижение денежного предложения._

Исходите из предположений о «недолете» валютного курса. 

_Как при этом изменится валютный курс в краткосрочной и долгосрочной перспективе?_

Недолет - это когда краткосрочная реакция курса на шок оказывается меньше (по модулю, но в том же направлении) чем долгосрочная. 

Начинаем решать с конца, чтобы понять как изменятся ожидания. Меньше денег навсегда означает, что в LR цены снизятся (т.к. $M/P=L(Y,i)$, а правая часть в LR константа). Снижение цен означает, что домашняя валюта укрепится номинально (т.к. $SP^* = P$ при неизменном $P^*$). То есть агенты будут ожидать укрепления курса в LR, соответственно $S^e$ снизится.

Кривая UIP уже в SR сдвинется вниз-влево из-за снижения параметра $S^e$

На нижней части диаграммы в SR кривая $M/P$ сдвинется вверх из-за снижения М. При этом предположения о «недолете» позволяют сделать вывод, что эффекты от изменения Y в SR оказывают значительное влияние на кривую md, и она сдвигается влево-вверх.
Итого, в SR равновесии ставка задается пересечением нового $M^d$ и нового $M/P$, и проецируется на новую UIP. 

В LR равновесии цены снизятся так, чтобы ставка вернулась к начальной. Зеленые линии на диаграмме.
