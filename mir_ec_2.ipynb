## 2.  Валютный курс

- Прямой валютный курс - стоимость единицы нац. валюты в иностранной валюты
- Обратный валютный курс - стоимость единицы иностранной валюты в нац. валюты

Если режим фиксированного (квази-фиксированного) валютного курса:

- Ревальвация нац. валюты - повышение стоимости нац. валюты в терминах иностранной валюты
- Девальвация нац. нац. валюты - снижение стоимости нац. валюты в терминах иностранной валюты

_Для того чтобы решить поддерживается ли в экономике фиксированный (квази-фиксированный) или плавающий валютный курс, не достаточно посмотреть официальные заявления ЦБ_: ЦБ не всегда объявляет правду о своем целеполагании. Есть много случаев контролируемого плавания, где ЦБ для себя определяет коридор или целевые показатели курса, но не объявляет о них рынку. Или же ЦБ заявляет о плавании, но иногда все-таки вмешивается в валютный рынок. 

### Структура рынка

Различают три сегмента валютного рынка для валюты каждой страны: 
- Внутренний валютный рынок (on-shore): операции резидентов с резидентам
- Внешний валютный рынок (off-shore): операции нерезидентов с нерезидентам 
- Трансграничные операции с национальной валютой: операции нерезидентов с резидентам


Для участников валютного рынка существует валютный риск: откуда 
потребность в его хеджировании, с помощью форвардных валютных контрактов 

- Спотовый (текущий) валютный курс: курс покупки/продажи валюты немедленно (в тот же день)
- Форвардный валютный курс (F) – заранее оговоренный курс покупки/продажи валюты в определенный момент в будущем (через 1-12 месяцев)

_Спекулянты на валютном (и в целом финансовом) рынке не всегда получают прибыль_:  У транзакций на финансовом рынке всегда 2 главные стороны, если не считать посредника. И, как правило, в итоге ожидания не совпадают с реальностью. Но надо различать прибыль в смысле разницы между ожидаемой и фактической прибылью транзакции, и прибыль в смысле разницы наличия возможности для транзакции и полного отсутствия такой возможности. В первом случае всегда одна сторона выигрывает, а вторая проигрывает. Во втором случае вполне могут выиграть обе стороны, в этом и состоит выигрыш от торговли.


Валютный своп – сочетание двух сделок (спотовая + форвардная или форвардная  + форвардная), обмен одной валюты на другую по оговоренному курсу  с последующим обратным обменом валют по оговоренному форвардному курсу.


_Операция «валютный своп» аналогична операции «репо», где в качестве залога выступают не ценные бумаги, а другая валюта_ (используется для управления ликвидностью )


### Реальный валютный курс

- Укрепление национальной валюты (appreciation) - нац. валюта становится дороже, т.е. снижение обратного валютного курса (повышение прямого валютного курса)

- Ослабление национальной валюты (depreciation) - нац. валюта становится дешевле, т.е. повышение обратного валютного курса (снижение прямого валютного курса)


Реальный валютный курс (Real exchange rate) - относительная стоимость корзины иностранных товаров в терминах отечественных товаров: $$RER_t = q_t = \dfrac{S_t P^{*}_t}{P_t}$$


- Укрепление нац. валюты в реальном выражении - увеличение стоимости корзины отечественных товаров относительно стоимости корзины иностранных товаров, $RER_t \downarrow$:

$$ \dot{RER} = \dot{S} - \pi^* - \pi < 0$$

- Ослабление нац. валюты в реальном выражении - уменьшение стоимости корзины отечественных товаров относительно стоимости корзины иностранных товаров 
- увеличивается), $RER_t \uparrow$:

$$ \dot{RER} = \dot{S} - \pi^* - \pi > 0$$

«Внутренний» $ RER = \frac{P_T}{ P_N} $ - мера внутренней
конкурентоспособности торгуемых и неторгуемых товаров
в производстве (конкуренция за ресурсы – факторы
производства) и в потреблении. Укрепление $ RER \Rightarrow $ неторгуемые товары становятся
относительно более дорогими, чем торгуемые




##### Задача

_На 2012 курс иены к доллару равен 76,94 иен за долл. За 2012 год цены  в США выросли на 1,7%, а цены в Японии снизились на 0,1%. Если на 2013 г. курс иены к доллару составил 86,67 иен за долл., то на начало 2013 г. иена укрепилась или ослабла в реальном выражении? _

Пусть Япония домашняя страна. Пусть стартовые уровни цен $P12$ и $P12^*$. Тогда 

$$RER12 = S12 \frac{P12^*}{P12} = 76.94 \frac{P12^*}{P12}$$

Если инфляция $pi$, то уровень цен через год $Pt+1= Pt(1+pi)$

$$RER13 = S13 \frac{P13^*}{P13} = 86.67 \frac{P12^* (1+0.017)}{P12(1-0.001} = 86.67*101.7/99.9 \frac{P12^*}{P12} =  88.23 \frac{P12^*}{P12}  > 76.94 \frac{P12^*}{P12} = RER12 $$

Реальный курс вырос, то есть иена ослабла в реальном выражении.

_Можно ли что-то сказать про переоцененность или недооцененность иены согласно ППС?_

Наши псевдо-RERы не показывают уровень, так как неизвестны настоящие стартовые уровни цен $P12$ и $P12^*$. Так что про ППС (сравнение RER с 1) ничего нельзя сказать.


## ППС

Закон одной цены (LOOP) - в долгосрочном периоде цены должны исключать арбитраж в международной торговле, для каждого товара/услуги, торгуемой на мировом рынке, нужно равенство:
$$ P_i = EP^*_i $$

Гипотеза абсолютного паритета покупательной 
способности (PPP) – обменный курс между валютами двух стран должен
уравнивать стоимость репрезентативных корзин, т.е. все уровни цен должны быть одинаковы, если исчислены в одной валюте: 
$$ P = EP^* $$ 

Относительный паритет: $ P = \alpha EP^* $

Объяснения отклонений от паритетов:
- Транспортные и транзакционные издержки, которые в частности могут меняться со временем
- Товары могут быть неоднородными (differentiated products), способность производителей дискриминировать между рынками разных стран 
- Существование неторгуемых товаров (цены не обязаны сравняться, если арбитраж невозможен, теория Балассы-Самуэльссона предсказывает постепенное удорожание реального курса по мере развития экономики страны)
- Монополии и олигополии на рынках товаров
- Технические причины (разные потребительские корзины)
- «Жесткость» цен товаров и услуг

Real effective exchange rate: $$ REER = \prod_{i=1}^N \left( \dfrac {S_i P^{*}_i}{P_i} \right)^{ w_i} $$

Nominal effective exchange: $$ NEER = \prod_{i=1}^N S_i^{ w_i}$$

$w_i$ - вес страны i во внешней торговле (экспорте и
импорте) домашней страны

Выбор ценового индекса:
- ИПЦ (CPI) – цены потребительской корзины, включает также импортные товары
- ИЦП (PPI – цены производителей)  и дефлятор ВВП (цены корзины производства экономики )  включает производимые товары, которые могут мало потребляться внутри экономики 
- индексы цен торгуемых и неторгуемых товаров 


### Долгосрочный (равновесный) RER

Граница производственных возможностей (Production
Possibility Frontier – PPF) – максимально возможная
комбинация производства двух товаров при условии
использования всех ресурсов.

RER задает бюджетное ограничение: $$ GDP=Y_T P_T +Y_N P_N \\ tg(\alpha) = - \frac{P_T}{P_N} $$

Оптимальная комбинация производства $Y_T$ и $Y_N$,
обеспечивающая максимальный ВВП - точка касания PPF и б.о.

Если RER ослабевает (торгуемые товары становятся
относительно более дорогими), то наклон бюджетного
ограничения увеличивается и становится более
выгодно производить больше торгуемых товаров


__ERER – equilibrium RER__ - определяют технологии (PPF), ресурсы (PPF)
и предпочтения потребителей (кривые безразличия).

ERER – RER, согласующийся с
внутренним равновесием ($Y_N = С_N$)
и внешним равновесием ($Y_T = С_T$) 

Эффект Баласса-Самуэльсона (BS)  -  существует положительная зависимость
между уровнями цен и доходами на душу населения в
странах, выраженных в одной и той же валюте

Причины: наличие в экономиках неторгуемых товаров и
разный уровень производительности сектора торгуемых
товаров в богатых и бедных странах


Основные предпосылки в модели Баласса-Самуэльсона (BS):
-  в обеих экономиках единственный фактор это труд, на совершенно конкурентном рынке внутри страны, но без возможности переехать в другую страну
-  в обеих экономиках производительность труда в торгуемом секторе больше в развитых странах, производительности труда в неторгуемых секторах равны
- фирмы работают в совершенной конкуренции, то есть зарплата равна предельной производительности





##### Задача

_Пусть общий уровень цен  в домашней $P$ и иностранной экономиках $P^*$  представляет собой среднегеометрическое цен торгуемого и неторгуемого товаров с одинаковым весом торгуемого товара a:_

$$ P = {P_T^a}{P_N^{(1-a)}},  P^*= (P_T^* )^a (P_N^* )^{(1-a )}$$ 


_ППС выполняется для сектора торгуемых товаров:_ 

$$P_T=SP_T^*, S – валютный \ курс $$

_Получите выражение для «внешнего» реального валютного курса (отношения стоимости потребительских корзин товаров и услуг в двух странах, выраженных в одной валюте) через цены торгуемых и неторгуемых товаров в двух экономиках. Будет ли выполняться ППС для корзин товаров?_


$$RER = S \frac{P^*}{P} = S \frac{  (P_T^* )^a (P_N^* )^{(1-a )} }{ {P_T^a}{P_N^{(1-a)}}}$$

По закону одной цены $S P_T^* = P_T$, так что кое-что в формуле выше сокращается: 

$$ RER =    ( \frac{  S P_N^* }{P_N})^{(1-a )} $$

ППС для корзины не обязан выполняться, так как для неторгуемых товаров невозможен арбитраж, а значит и закон одной цены не действует


_Используя предпосылки модели BS, покажите, что общий уровень цен в богатой стране будет выше, чем в бедной_ 

Запишем равенство производительностей в неторгуемых секторах стран и выразим через него отношение цен в неторгуемых секторах:

$$ W = P_N MPL_N,  W^*= P_N^* MPL_N^* $$ 

отсюда из равенства неторгуемых производительностей

$$ \frac{W}{P_N} = \frac{W^*}{P_N^*} $$ 

$$ \frac{P_N^*}{P_N} = \frac{W^*}{W}  $$

Выразим зарплаты через торгуемые производительности:

$$ W = P_T MPL_T , W^*= P_T^*MPL_T^* $$


$$ \frac{P_N^*}{P_N} = \frac{ P_T^*MPL_T^*}{P_T MPL_T }  $$


подставим закон одной цены для $P_T, P_T^* $


$$ \frac{P_N^*}{P_N} = \frac{ MPL_T^*}{S \ MPL_T }  $$


то есть  

$$ S \frac{P_N^*}{P_N} = \frac{ MPL_T^*}{ \ MPL_T }  $$


Если богатая страна *, то торгуемая производительность там выше и $$ S \frac{P_N^*}{P_N}   >1 $$

То есть цены на неторгуемые товары в * тоже выше. 

Значит и общий уровень цен выше, т.к. торгуемые товары стоят одинаково.

Ну и конечно это как раз RER и есть (с точностью до степени). То есть если торгуемая производительность в * больше (развитая страна), то RER для домашней страны (развивающаяся) обесценен по сравнению с ППС.

### Модель спроса-предложения нац. товаров

RER определяется из пересечения относительных
спроса и предложения на наши и иностранные товары

$RS = \frac{Y}{Y^*}, RD(\frac{Y}{Y^*},\frac{P^*}{P})$

- Если растет спрос на наши товары: RD вправо,  RER укрепляется
- Если растет наш выпуск: RS вправо, RER слабеет. Интуитивно: наших товаров больше, значит они дешевле

##### Задача

_Развивающиеся страны в среднем растут быстрее развитых. Что этот факт может сказать нам о возможной динамике реального обменного курса развивающихся стран в рамках модели Баласса-Самуэльссона?_

Быстрый рост видимо сопровождается технологической сходимостью. Значит, производительность развивающихся стран в торгуемых товарах растет, а разрыв с развитыми странами сокращается. По модели BS это приведет к росту уровня цен в развивающихся странах и соответственно к укреплению их валют в реальном выражении.

_А в рамках модели относительного спроса-предложения?_

А в модели RS/RD опережающий рост ВВП приводит к повышению относительного предложения товаров развивающихся стран и соответственно к снижению относительных цен на них. То есть развивающийся RER обесценивается. 
BS более похожа на реальность.

### Простая модель открытой экономики (ПМОЭ)

В открытой экономике можно давать или брать в долг у остального мира по
заданной извне процентной ставке $r$

$$ C_1 + S_1 = Y_1 \\  C_2 = Y_2 + S_2 = Y_2 +  (1+r)S_1  = Y_2 + (1+r)(Y_1 -C_1 )$$ 

Откуда  получаем  уравнение прямой бюджетного ограничения  для задачи потребителя:

$$ \max U(C_1,C_2 ) \\ s.t.: (1+r)C_1 + С_2 = (1+r)Y_1 + Y_2 $$





### ПМОЭ: счет текущих операций

Сбережения $S_1$ это счет текущих операций $CA_1$
- инвестиций в 1м периоде нет, $CA = S - I$ 

Выгоды от открытости: 
- Можно сместиться от точки автаркии
- Полезность выше для любого $ r \neq r_A $  
    - Если ставка ниже чем в автаркии, то в 1м периоде импортируем
    - Если ставка выше чем в автаркии, то в 1м периоде экспортируем
  

### ПМОЭ:  реальная ставка

- Для сбережений предполагаем, что  «эффект замещения» доминирует, то есть  $S'(r) > 0$ 
- Для инвестиции при росте ставки труднее занимать капитал, то есть $I'(r) < 0$ 
- $CA'(r) > 0$

Малая экономика:
- Мировая ставка – внешний параметр
- Малая экономика – price taker 
- Изменения в нашей экономике могут сдвигать кривую CA(r), но не влияют на ставку

Большая экономика:
- Изменения в нашей экономике меняют мировую ставку: $CA(r) + CA_{ROW} (r) = 0$


### ПМОЭ: алгебра 

$$ U(C_1,C_2) = \ln(C_1) + \beta \ln(C_2 ) \\ s.t.: C_1 + \frac{C_2}{1+r} = Y_1 + \frac{Y_2}{1+r} $$

$NPV$ потребления = $NPV$ выпуска (эндоумента)

Условия 1-го порядка (уравнение Эйлера):

$$ \frac{\beta  u'(C_2) }{u'(C_1)} = \frac{1}{1+r} $$

В т. максимума потребитель не может увеличить полезность за счет перераспределения потребления между периодами.  Предельная степень заменимости сегодняшнего потребления на завтрашнее потребление = цена единицы потребления завтра в терминах потребления сегодня

- Домашние потребители менее терпеливы, чем остальной мир (ROW): $$ \beta  < \frac{1}{1+r} \Rightarrow C_1>C_2 $$
- Домашние потребители более терпеливы, чем ROW:  $$ \beta  > \frac{1}{1+r} \Rightarrow C_1<C_2 $$
- Домашние потребители также терпеливы, как ROW (полное сглаживание потребления):  $$ \beta  =  \frac{1}{1+r} \Rightarrow C_1=C_2 \\ Y_1 - C_1 = Y_2 - C_2 + rS_1  \\ CA_1 = - CA_2$$

### Глобальные дисбалансы

Глобальные дисбалансы – наличие продолжительных и
значительных профицитов/дефицитов СТО в крупных странах

Др. словами на протяжении длительного времени (десятилетий)
одни страны постоянно чистые кредиторы, а другие – чистые заемщики 

Cumulative  current  account: $$ CCA = \sum_{\tau = t_0}^{t} CA_{\tau}  $$

Чистая международная инвестиционная позиция (Net
International Investment Position): 
 
$$ NIIP = CCA + переоценка\ активов$$ 

##### Задача:

Налоги забираются из эндаумента перед тем как потребитель делает решение о потреблении. Соответственно потребитель выбирает уровень сбережений в первом периоде в виде:
$$S1 = Y1-T1-C1$$
$$C2 = Y2-T2 + (1+r)S1$$

БО:  $Y2-T2-С2+(1+r)(Y1-T1-C1) = 0 $

Так как потребитель никак не влияет на $Ti$ и $Gi$, то он воспринимает их как константы, и по сути задача потребителя это максимизация $u(C1)+ \beta u(C2)$ при эндаументе $Y1-T1, Y2-T2$, то есть классическая форма модели межвременного выбора. 

Заметим, что тут предполагается, что потребитель ведет себя как ведомый по Штакельбергу. Это резонное предположение для групп людей («от меня ничего не зависит»).


На графике в пространстве $С1,С2$ линия бюджетного ограничения будет проходить через точку $Y1-T1, Y2-T2$ с наклоном -$(1+r)$ – множителем при $С1$

Решение задачи потребителя в общем виде:

Лагранжиан задачи максимизации: $L = u(C1)+ \beta u(C2) + \lambda (Y2-T2-С2+(1+r)(Y1-T1-C1))$ 

Условия первого порядка: $\partial L/ \partial  Ci = 0$, то есть $u’(C1)-\lambda(1+r) = 0$ и $\beta u’(C2)- \lambda =0$ 

Исключая $\lambda$, получаем уравнение Эйлера $u’(C1)/u’(C2)= \beta (1+r) $

Если $\beta=1/(1+r)$ то $u’(C1)=u’(C2)$ и значит $С1=С2$, потребление полностью сглаживается

__СТО в 1-м (CA1) и 2-м (CA2) периодах__ 


$$CA = S-I$$

Инвестиции нулевые по условию. Значит $САi = Si$

Сначала вычислим сбережения (доход минус съеденное) в каждом из периодов: 

$$S = ВНП–C–G$$

Заметим, что $ВНПi$ далеко не всегда совпадает с $Yi$, так как $ ВНП=ВВП + NFP $  и во 2м периоде это важно.


Сбережения в 1м периоде: $$S1 = ВНП-C1-G1 = Y1-T1-C1$$
Сбережения во 2м периоде: $$S2 = ВНП-С2-G2 = Y2-T2-C2 + rS1$$

Подставляем $Y2-T2-C2$ из бюджетного ограничения и получаем

$$S2 = -(1+r)(Y1-T1-C1) +rS1 = -(1+r)S1 +rS1 = -S1 $$

В результате $CA2 = -CA1$






__Если экономика изолирована от остального мира: __

В автаркии $Ci = Yi-Ti$, так как больше товары брать неоткуда, и некуда отдавать. 

Соответственно подставляем $Ci$ в уравнение Эйлера и получаем 

$$r = u’(Y1-G1)/\beta u’(Y2-G2) -1$$

Заметим, что реальная ставка ничем не ограничена, может быть как положительной, так и отрицательной. Отрицательную ставку интуитивно можно воспринимать как плату за перенос хоть какой-то части товара во 2й период. Если много желающих перенести, то этот фактор перевесит желание агентов  пользоваться товарами в 1м периоде, которое повышает рыночную ставку.


__Пусть $\beta =1/(1+r)$  и изначально ожидалось, что $Y1-G1=Y2-G2$, но затем происходит рост госрасходов в 1-м периоде на $\Delta G$__ 


Если  $\beta =1/(1+r)$,  С1 = С2 при любых движениях эндаумента. 

Изначально $Y1-G1 = Y2-G2 = С1 = С2 $, что значит нулевые сбережения в 1м периоде и $CAi = 0$

Если G1 растет, это уменьшает располагаемый доход частного сектора, а значит, будет снижаться и уровень потребления в обоих периодах $C1=C2$. 
Значит, во 2м периоде становится $Y2-G2>C2$ и соответственно $CA2>0$. Отсюда следует, что $CA1<0$ и $С1$ падает меньше, чем растет $G1$.



__Пусть $\beta =1/(1+r)$  и изначально ожидалось, что $Y1-G1=Y2-G2$, но затем происходит рост госрасходов в 2-м периоде на $\Delta G$__ 

Как и в предыдущем пункте всегда $С1 = С2$. Располагаемый доход частного сектора падает только во 2м периоде, а вот потребление будет падать в обоих периодах. Это значит что $CA1>0, CA2<0$.


__Пусть $\beta =1/(1+r)$  и изначально ожидалось, что $Y1-G1=Y2-G2$, но затем происходит рост госрасходов и в 1-м, и во 2-м периоде на $\Delta G$__ 


Как и в предыдущем пункте С1=С2 всегда. $Y1-G1=Y2-G2 = С1= С2$  по-прежнему удовлетворяет условиям равновесия (Эйлер и бюджетное ограничение). Так что $Ci$ упадет ровно на $\Delta G$, а $CAi=0$  останутся.



__Если государство ставит целью максимизацию общественного благосостояния:__

Гос-во знает, что потребитель выбирает $Ci$, удовлетворяя уравнению Эйлера и бюджетному ограничению. 

Получаются оптимальные решения $С1^*(G1,G2),  С2^*(G1,G2)$.


Задача максимизации для государства: 

$$ \max u(C1^*(G1,G2)))+ \beta u(C2^*(G1,G2)) + v(G1)+ \beta v(G2) $$

БО  государства (оно не может использовать больше товаров, чем есть в экономике)  

$$ Y2-G2+(1+r)(Y1-G1) > 0 $$
 
Если БО превратится в равенство, то это будет означать что для потребителей останется ровно одна доступная точка $С1=С2=0$. Вряд ли это когда-либо окажется оптимальным решением. Обычно в задачах, где есть потребление, мы по умолчанию предполагаем, что нулевое потребление невозможно, агент «умрет с голоду». В этой задаче это предположение неважно, но в целом оно часто используется.


## 2.  Валютный курс

- Прямой валютный курс - стоимость единицы нац. валюты в иностранной валюты
- Обратный валютный курс - стоимость единицы иностранной валюты в нац. валюты

Если режим фиксированного (квази-фиксированного) валютного курса:

- Ревальвация нац. валюты - повышение стоимости нац. валюты в терминах иностранной валюты
- Девальвация нац. нац. валюты - снижение стоимости нац. валюты в терминах иностранной валюты

_Для того чтобы решить поддерживается ли в экономике фиксированный (квази-фиксированный) или плавающий валютный курс, не достаточно посмотреть официальные заявления ЦБ_: ЦБ не всегда объявляет правду о своем целеполагании. Есть много случаев контролируемого плавания, где ЦБ для себя определяет коридор или целевые показатели курса, но не объявляет о них рынку. Или же ЦБ заявляет о плавании, но иногда все-таки вмешивается в валютный рынок. 

### Структура рынка

Различают три сегмента валютного рынка для валюты каждой страны: 
- Внутренний валютный рынок (on-shore): операции резидентов с резидентам
- Внешний валютный рынок (off-shore): операции нерезидентов с нерезидентам 
- Трансграничные операции с национальной валютой: операции нерезидентов с резидентам


Для участников валютного рынка существует валютный риск: откуда 
потребность в его хеджировании, с помощью форвардных валютных контрактов 

- Спотовый (текущий) валютный курс: курс покупки/продажи валюты немедленно (в тот же день)
- Форвардный валютный курс (F) – заранее оговоренный курс покупки/продажи валюты в определенный момент в будущем (через 1-12 месяцев)

_Спекулянты на валютном (и в целом финансовом) рынке не всегда получают прибыль_:  У транзакций на финансовом рынке всегда 2 главные стороны, если не считать посредника. И, как правило, в итоге ожидания не совпадают с реальностью. Но надо различать прибыль в смысле разницы между ожидаемой и фактической прибылью транзакции, и прибыль в смысле разницы наличия возможности для транзакции и полного отсутствия такой возможности. В первом случае всегда одна сторона выигрывает, а вторая проигрывает. Во втором случае вполне могут выиграть обе стороны, в этом и состоит выигрыш от торговли.


Валютный своп – сочетание двух сделок (спотовая + форвардная или форвардная  + форвардная), обмен одной валюты на другую по оговоренному курсу  с последующим обратным обменом валют по оговоренному форвардному курсу.


_Операция «валютный своп» аналогична операции «репо», где в качестве залога выступают не ценные бумаги, а другая валюта_ (используется для управления ликвидностью )


### Реальный валютный курс

- Укрепление национальной валюты (appreciation) - нац. валюта становится дороже, т.е. снижение обратного валютного курса (повышение прямого валютного курса)

- Ослабление национальной валюты (depreciation) - нац. валюта становится дешевле, т.е. повышение обратного валютного курса (снижение прямого валютного курса)


Реальный валютный курс (Real exchange rate) - относительная стоимость корзины иностранных товаров в терминах отечественных товаров: $$RER_t = q_t = \dfrac{S_t P^{*}_t}{P_t}$$


- Укрепление нац. валюты в реальном выражении - увеличение стоимости корзины отечественных товаров относительно стоимости корзины иностранных товаров, $RER_t \downarrow$:

$$ \dot{RER} = \dot{S} - \pi^* - \pi < 0$$

- Ослабление нац. валюты в реальном выражении - уменьшение стоимости корзины отечественных товаров относительно стоимости корзины иностранных товаров 
- увеличивается), $RER_t \uparrow$:

$$ \dot{RER} = \dot{S} - \pi^* - \pi > 0$$

«Внутренний» $ RER = \frac{P_T}{ P_N} $ - мера внутренней
конкурентоспособности торгуемых и неторгуемых товаров
в производстве (конкуренция за ресурсы – факторы
производства) и в потреблении. Укрепление $ RER \Rightarrow $ неторгуемые товары становятся
относительно более дорогими, чем торгуемые




##### Задача

_На 2012 курс иены к доллару равен 76,94 иен за долл. За 2012 год цены  в США выросли на 1,7%, а цены в Японии снизились на 0,1%. Если на 2013 г. курс иены к доллару составил 86,67 иен за долл., то на начало 2013 г. иена укрепилась или ослабла в реальном выражении? _

Пусть Япония домашняя страна. Пусть стартовые уровни цен $P12$ и $P12^*$. Тогда 

$$RER12 = S12 \frac{P12^*}{P12} = 76.94 \frac{P12^*}{P12}$$

Если инфляция $pi$, то уровень цен через год $Pt+1= Pt(1+pi)$

$$RER13 = S13 \frac{P13^*}{P13} = 86.67 \frac{P12^* (1+0.017)}{P12(1-0.001} = 86.67*101.7/99.9 \frac{P12^*}{P12} =  88.23 \frac{P12^*}{P12}  > 76.94 \frac{P12^*}{P12} = RER12 $$

Реальный курс вырос, то есть иена ослабла в реальном выражении.

_Можно ли что-то сказать про переоцененность или недооцененность иены согласно ППС?_

Наши псевдо-RERы не показывают уровень, так как неизвестны настоящие стартовые уровни цен $P12$ и $P12^*$. Так что про ППС (сравнение RER с 1) ничего нельзя сказать.


## ППС

Закон одной цены (LOOP) - в долгосрочном периоде цены должны исключать арбитраж в международной торговле, для каждого товара/услуги, торгуемой на мировом рынке, нужно равенство:
$$ P_i = EP^*_i $$

Гипотеза абсолютного паритета покупательной 
способности (PPP) – обменный курс между валютами двух стран должен
уравнивать стоимость репрезентативных корзин, т.е. все уровни цен должны быть одинаковы, если исчислены в одной валюте: 
$$ P = EP^* $$ 

Относительный паритет: $ P = \alpha EP^* $

Объяснения отклонений от паритетов:
- Транспортные и транзакционные издержки, которые в частности могут меняться со временем
- Товары могут быть неоднородными (differentiated products), способность производителей дискриминировать между рынками разных стран 
- Существование неторгуемых товаров (цены не обязаны сравняться, если арбитраж невозможен, теория Балассы-Самуэльссона предсказывает постепенное удорожание реального курса по мере развития экономики страны)
- Монополии и олигополии на рынках товаров
- Технические причины (разные потребительские корзины)
- «Жесткость» цен товаров и услуг

Real effective exchange rate: $$ REER = \prod_{i=1}^N \left( \dfrac {S_i P^{*}_i}{P_i} \right)^{ w_i} $$

Nominal effective exchange: $$ NEER = \prod_{i=1}^N S_i^{ w_i}$$

$w_i$ - вес страны i во внешней торговле (экспорте и
импорте) домашней страны

Выбор ценового индекса:
- ИПЦ (CPI) – цены потребительской корзины, включает также импортные товары
- ИЦП (PPI – цены производителей)  и дефлятор ВВП (цены корзины производства экономики )  включает производимые товары, которые могут мало потребляться внутри экономики 
- индексы цен торгуемых и неторгуемых товаров 


### Долгосрочный (равновесный) RER

Граница производственных возможностей (Production
Possibility Frontier – PPF) – максимально возможная
комбинация производства двух товаров при условии
использования всех ресурсов.

RER задает бюджетное ограничение: $$ GDP=Y_T P_T +Y_N P_N \\ tg(\alpha) = - \frac{P_T}{P_N} $$

Оптимальная комбинация производства $Y_T$ и $Y_N$,
обеспечивающая максимальный ВВП - точка касания PPF и б.о.

Если RER ослабевает (торгуемые товары становятся
относительно более дорогими), то наклон бюджетного
ограничения увеличивается и становится более
выгодно производить больше торгуемых товаров


__ERER – equilibrium RER__ - определяют технологии (PPF), ресурсы (PPF)
и предпочтения потребителей (кривые безразличия).

ERER – RER, согласующийся с
внутренним равновесием ($Y_N = С_N$)
и внешним равновесием ($Y_T = С_T$) 

Эффект Баласса-Самуэльсона (BS)  -  существует положительная зависимость
между уровнями цен и доходами на душу населения в
странах, выраженных в одной и той же валюте

Причины: наличие в экономиках неторгуемых товаров и
разный уровень производительности сектора торгуемых
товаров в богатых и бедных странах


Основные предпосылки в модели Баласса-Самуэльсона (BS):
-  в обеих экономиках единственный фактор это труд, на совершенно конкурентном рынке внутри страны, но без возможности переехать в другую страну
-  в обеих экономиках производительность труда в торгуемом секторе больше в развитых странах, производительности труда в неторгуемых секторах равны
- фирмы работают в совершенной конкуренции, то есть зарплата равна предельной производительности





##### Задача

_Пусть общий уровень цен  в домашней $P$ и иностранной экономиках $P^*$  представляет собой среднегеометрическое цен торгуемого и неторгуемого товаров с одинаковым весом торгуемого товара a:_

$$ P = {P_T^a}{P_N^{(1-a)}},  P^*= (P_T^* )^a (P_N^* )^{(1-a )}$$ 


_ППС выполняется для сектора торгуемых товаров:_ 

$$P_T=SP_T^*, S – валютный \ курс $$

_Получите выражение для «внешнего» реального валютного курса (отношения стоимости потребительских корзин товаров и услуг в двух странах, выраженных в одной валюте) через цены торгуемых и неторгуемых товаров в двух экономиках. Будет ли выполняться ППС для корзин товаров?_


$$RER = S \frac{P^*}{P} = S \frac{  (P_T^* )^a (P_N^* )^{(1-a )} }{ {P_T^a}{P_N^{(1-a)}}}$$

По закону одной цены $S P_T^* = P_T$, так что кое-что в формуле выше сокращается: 

$$ RER =    ( \frac{  S P_N^* }{P_N})^{(1-a )} $$

ППС для корзины не обязан выполняться, так как для неторгуемых товаров невозможен арбитраж, а значит и закон одной цены не действует


_Используя предпосылки модели BS, покажите, что общий уровень цен в богатой стране будет выше, чем в бедной_ 

Запишем равенство производительностей в неторгуемых секторах стран и выразим через него отношение цен в неторгуемых секторах:

$$ W = P_N MPL_N,  W^*= P_N^* MPL_N^* $$ 

отсюда из равенства неторгуемых производительностей

$$ \frac{W}{P_N} = \frac{W^*}{P_N^*} $$ 

$$ \frac{P_N^*}{P_N} = \frac{W^*}{W}  $$

Выразим зарплаты через торгуемые производительности:

$$ W = P_T MPL_T , W^*= P_T^*MPL_T^* $$


$$ \frac{P_N^*}{P_N} = \frac{ P_T^*MPL_T^*}{P_T MPL_T }  $$


подставим закон одной цены для $P_T, P_T^* $


$$ \frac{P_N^*}{P_N} = \frac{ MPL_T^*}{S \ MPL_T }  $$


то есть  

$$ S \frac{P_N^*}{P_N} = \frac{ MPL_T^*}{ \ MPL_T }  $$


Если богатая страна *, то торгуемая производительность там выше и $$ S \frac{P_N^*}{P_N}   >1 $$

То есть цены на неторгуемые товары в * тоже выше. 

Значит и общий уровень цен выше, т.к. торгуемые товары стоят одинаково.

Ну и конечно это как раз RER и есть (с точностью до степени). То есть если торгуемая производительность в * больше (развитая страна), то RER для домашней страны (развивающаяся) обесценен по сравнению с ППС.

### Модель спроса-предложения нац. товаров

RER определяется из пересечения относительных
спроса и предложения на наши и иностранные товары

$RS = \frac{Y}{Y^*}, RD(\frac{Y}{Y^*},\frac{P^*}{P})$

- Если растет спрос на наши товары: RD вправо,  RER укрепляется
- Если растет наш выпуск: RS вправо, RER слабеет. Интуитивно: наших товаров больше, значит они дешевле

##### Задача

_Развивающиеся страны в среднем растут быстрее развитых. Что этот факт может сказать нам о возможной динамике реального обменного курса развивающихся стран в рамках модели Баласса-Самуэльссона?_

Быстрый рост видимо сопровождается технологической сходимостью. Значит, производительность развивающихся стран в торгуемых товарах растет, а разрыв с развитыми странами сокращается. По модели BS это приведет к росту уровня цен в развивающихся странах и соответственно к укреплению их валют в реальном выражении.

_А в рамках модели относительного спроса-предложения?_

А в модели RS/RD опережающий рост ВВП приводит к повышению относительного предложения товаров развивающихся стран и соответственно к снижению относительных цен на них. То есть развивающийся RER обесценивается. 
BS более похожа на реальность.

### Режимы валютного курса


Зачем вмешиваться в валютные рынки
- Стабильность и предсказуемость обменного курса являются основным фактором развития прямых иностранных инвестиций и международной торговли
- Фиксированный курс «импортирует» денежную дисциплину
- Курсовые колебания влияют на инфляцию, поскольку импортируемые товары являются частью корзины потребления

Режимы ER: 
- чисто плавающий режим 
- политика обменного курса – когда ЦБ вмешивается в валютный рынок на систематической основе

Важно, что ЦБ говорит в отношении политики ER:
- ЦБ может активно вмешиваться в валютные рынки, не раскрывая свои реальные цели. Это случай т.н. дискреционного вмешательства
- ЦБ может заявить, что он будет вмешиваться столько, сколько необходимо, чтобы зафиксировать уровень обменного курса против некоторой иностранной валюты
