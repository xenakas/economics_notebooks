### Partie  3.1:


<b>Grammaire : les temps du passé et l'accord du participe passé
</b>

<table border="0">
 <tr>
    <td>L'imparfait (IMP) :</td>
    <td>The imperfect:</td>
 </tr>
 <tr>
    <td> Formation de l'imparfait :<br /><br />
    - radical du verbe conjugué avec nous au présent + les terminaisons
      (ais/ais/ait/ions/iez/aient)<br /><br />

      nous prenons ==> je prenais<br /><br />
      nous finissons ==>je finissais<br /><br />
      Exception : être<br /><br /> j'étais, tu étais, il/elle/on était, nous étions,vous étiez, ils étaient. <br /><br />
        
        - exprime une habitude dans le passé :
        
 "Ma mère me donnait chaque jour un carré de chocolat au lait."
      <br /><br />
      
    - présente une personne, un lieu, un objet dans un récit :
   
      "Il portait un jean et un teeshirt blanc. Il souriait !" <br /><br />
    - présente des actions et des états sans limites de temps précises :
    
      "Quand nous étions petits, nous voulions former un groupe."
<br /><br />


</td>
    <td>     Formation of the imperfect:
  
    <br /><br />
    - verb radical conjugated with us in the present + the terminations
        (ais/ais/ait/ions/iez/aient)<br /><br />


       we take ==>  I took<br /><br />
       we finish ==>  I finish<br /><br />
       Exception: to be<br /><br /> I was, you were, he / she / we were, we were, you were, they were
 <br /><br />
  - expresses a habit in the past:
  
       "My mother gave me a square of milk chocolate every day."<br /><br />
     - presents a person, a place, an object in a story:
     
       "He was wearing jeans and a white t-shirt, he was smiling!"<br /><br />
     - presents actions and states with unlimited time limits:
     
       "When we were little, we wanted to form a group."

<br /><br />

</td>
 </tr>
 
  <tr>
    <td>Le passé composé (PC) :</td>
    <td>The past composite (PC):</td>
 </tr>
 <tr>
    <td>

Formation du passé composé : <br /><br />

Le verbe conjugué avec l'auxiliaire "être" ou "avoir"<br /><br />

##### Tache 6

Les principales rubriques d'un CV:
Informations personnelles,  Compétences, Formation, Expériences professionnelles

Les éléments qui doivent figurer dans une lettre de motivation: Votre adresse, 	L’adresse du destinataire, 	La date et le lieu, 	L’objet de la lettre,	Le contenu, 	La signature

Les types des formules de politesse:

- Les formules amicale: Gros bisous, Amitiés, A bientôt

- Les formules soutenue: Veuillez agréer, Monsieur, mes salutations distinguées. Veuillez accepter, Monsieur l’expression de ma plus profonde gratitude. 

- Les formules modérée: 
Bien cordialement, Sincèrement
