#### grammar

L'imparfait (IMP) :
    - exprime une habitude dans le passé :
      "Ma mère me donnait chaque jour un carré de chocolat au lait."
    - présente une personne, un lieu, un objet dans un récit :
      "Il portait un jean et un teeshirt blanc. Il souriait !"
    - présente des actions et des états sans limites de temps précises :
      "Quand nous étions petits, nous voulions former un groupe."

Comment se forme l'imparfait ?

Formation de l'imparfait :
    - radical du verbe conjugué avec nous au présent + les terminaisons
      (ais/ais/ait/ions/iez/aient)
    - Exemple :
      nous prenons ==> prenons ==>je prenais
      nous finissons ==>finissons ==>je finissais

Le passé composé (PC) :
- présente des actions ou des états finis à un moment précis :
"Hier soir, nous sommes allés au cinéma."
- présente une succession d'événements dans un récit :
"Soudain, il est sorti de son véhicule et a tiré en l'air..."

Accord du participe passé :

- Pour les 14 verbes conjugué avec "être", l'accord se fait avec le sujet.
"Hier, vous êtes arrivés à quelle heure ?"

- Pour les verbes conjugués avec "avoir", l'accord se fait avec le COD si celui-ci
est placé devant le verbe.
"Rends-moi les films que je t'ai prêtés si tu les as vus."
[que et les (=films) sont COD de prêter et voir]


- Pour les verbes pronominaux, l'accord se fait avec le sujet si le verbe n'existe qu'à
la forme pronominale (s'absenter, s'évanouir,...) et si le verbe n'est pas réfléchi
(s'apercevoir, se taire, ...).
"Elle s'est absentée, je me suis tue."



- Pour les autres verbes pronominaux, l'accord se fait avec le COD (même règle qu'avec
"avoir").
"Il s'est lavé. Elles se sont lavées. Elle s'est lavé[Ø] les mains."
