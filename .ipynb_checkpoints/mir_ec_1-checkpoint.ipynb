# Мировая экономика

## 1. Система национальных счетов и платежный баланс

### Открытость экономики: товарные рынки

Экспорт товаров и услуг – Импорт товаров и услуг = Баланс товаров и услуг:

$$X–M=TB$$


Степень открытости товарного рынка в стране – экспорт и импорт товаров (и услуг), в % ВВП 

Малые открытые экономики – большая доля экспорта и импорта в ВВП, доля в мировом ВВП не большая

Закрытые экономики - низкая доля торговли в ВВП (США близка к таковым из-за удаленности и размера)

Экспорт и импорт могут быть больше, чем ВВП страны (пример – Сингапур), т.к. экспорт и импорт могут включать промежуточные товары, а ВВП – конечные товары, небольшие по размеру удачно расположенные (и с офшорным статусом) страны активно занимаются ре- экспортом

Торгуемые товары (экспортируемые, импортируемые) - конкурируют с иностранными товарами на внешних и внутренних рынках

При открытости товарных рынков выбор между отечественными и иностранными товарами влияет на выпуск в экономике:

спрос на отечественные товары↑ → выпуск (ВВП)↑

Главную роль в выборе между отечественными и иностранными товарами играют относительные цены на эти товары, посчитанные в единой валюте – реальный валютный курс

Открытость финансовых рынков позволяет:

- финансировать дефициты торговых балансов (точнее СТО) за счет притока иностранного капитала в страну, представляющего собой заимствования у иностранных экономических агентов

- сберегать в виде иностранных активов/давать в долг иностранным экономическим агентам образующиеся профициты торговых балансов (СТО) 

### Национальные счета

ВВП/GDP - стоимость всех конечных товаров и услуг, произведенных в экономике за определенный период времени (т.е. географический принцип):

$$ВВП=C+I+G+X−M$$


ВНП/GNP - стоимость всех конечных товаров и услуг, произведенных принадлежащими отечественным эконом. агентам (резидентам) факторами производства (трудом, капиталом) за определенный период времени (т.е. принцип собственности):

$$ВНП=ВВП+NFP$$


NFP - net factor payments from abroad

Доходы факторов производства за рубежом: 

- Зарплаты резидентов, работающих в др. странах
- Прибыль компаний от производства в др. странах

$$CA = X-M+NFP +NT $$

NT - баланс текущих трансфертов (net unilateral current transfers)

Если $NT=0$, то  $ВНП =C+I+G+CA$ 


### Платежный баланс

Платежный баланс (Balance of Payments, BoP) – отражает транзакции страны с остальным миром:

- Покупка-продажа товаров и услуг
- Покупка-продажа финансовых активов

Платежный баланс имеет 2 главных счета:
- Счет текущих операций (current account, CA)
- Финансовый счет (KA)

Как правило, транзакция с товаром (услугой) зеркально
отражается в транзакциях с финансовыми активами. В теории: $$CA=KA$$


В реальности:
- $CA<0$ хорошо, если покрывает временный спад выпуска или покрывает рост спроса на инвестиции, долги потом отдадим
- $CA<0$ плохо, если только поддерживает потребление, ведь не создается базы для возвращения долгов
- $CA>0$ хорошо, если временный рост выпуска – формируем резерв на будущее
- $CA>0$ плохо, если домашние инвестиции при этом падают, это сигнал о том что возможностей для внутренних инвестиций мало и резиденты предпочитают инвестировать в другие страны

### Финансовый счет (движения капитала)

$$ KA  = \Delta NFA = + \Delta (иностранные\ активы)- \Delta (иностранные \ обязательства)
$$

$NFA $ - Net foreign assets или изменение чистых
иностранных активов


$KA<0$ - чистый приток капитала, приток в страну иностранного капитала больше, чем отток из страны отечественного капитала заграницу $\rightarrow$ чистый внешний долг$\downarrow$

$KA>0$ - чистый отток капитала,  приток в страну иностранного капитала меньше, чем отток из страны отечественного капитала заграницу  $\rightarrow$ чистый внешний долг$\uparrow$



### Платежный баланс для страны нерезервной валюты

$$CA = KA^{к.р.} + \Delta RES$$

$KA$  кроме резервов -  это чистый приток-отток капитала по каналам частного сектора и госсектора, за
исключением операций ЦБ с международными резервами
(резервными активами)

$\Delta RES$ – изменение международных (валютных) резервов (резервных
активов)

- $CA-KA^{к.р.} < 0$ значит, что ЦБ за счет снижения международных резервов компенсирует дефицит валюты, поступающей в экономику по каналам торговли товарами и услугами (СA) и финансовыми активами (KA)
- $CA-KA^{к.р.} > 0$  значит, что ЦБ за счет увеличения международных резервов компенсирует избыток валюты, поступающей в экономику по каналам торговли товарами и услугами и финансовыми активами

### Платежный баланс для страны резервной валюты

$$ CA=KA^{к.р.} + \Delta RES_{Наш \ ЦБ}  - \Delta RES_{ ЦБ \ др.стран \ в\  валюте\  этой\  страны}$$

Как правило, у страны резервной валюты $ \Delta RES \approx 0 $

ЦБ др. стран покупают резервную валюту этой страны, что может служить основным или дополнительным
источников финансирования дефицита счета текущих
операций (кроме чистого притока инвестиций в
нерезервные финансовые активы, $KA^{к.р.} < 0$)

### Платежный баланс 

На практике:  

$$CA  +  Счет\ операций\ с\ капиталом + E\& O = KA
$$

Счет операций с капиталом:
- Продажа – покупка непроизведенных нефинансовых активов (патентов, лицензий)
- Чистое прощение иностранного долга (государственным и частным секторами)
- Гуманитарная помощь, безвозмездная передача и т.п.

E&O - Ошибки и пропуски (errors & omissions)
: не удается
правильно учесть все транзакции в статистике (часто ошибки
или намеренные искажения в статистике импорта, в оттоке
капитала, но ЦБ учатся их корректировать)
Таким образом:


###  Национальные сбережения

Национальные сбережения - то что мы произвели, но не потребили 
$$S = Y-C-G = I + CA $$
в закрытой экономике $CA=0$

$S_p = Y-T-C =  I_p - BB + CA$ -  сбережения частного сектора

- На финансирование частных инвестиций
- На финансирование дефицита гос. Бюджета
- На приобретение иностранных активов


$S_g = T – G$ -  сбережения государства

$BB = T – G – I_g$ -  баланс госбюджета (budget balance)




## 2. Intertemporal approach to CA

Межвременная тоговля – обмен ресурсами во времени

### Простая модель открытой экономики (ПМОЭ)

В открытой экономике можно давать или брать в долг у остального мира по
заданной извне процентной ставке $r$

$$ C_1 + S_1 = Y_1 \\  C_2 = Y_2 + S_2 = Y_2 +  (1+r)S_1  = Y_2 + (1+r)(Y_1 -C_1 )$$ 

Откуда  получаем  уравнение прямой бюджетного ограничения  для задачи потребителя:

$$ \max U(C_1,C_2 ) \\ s.t.: (1+r)C_1 + С_2 = (1+r)Y_1 + Y_2 $$





### ПМОЭ: счет текущих операций

Сбережения $S_1$ это счет текущих операций $CA_1$
- инвестиций в 1м периоде нет, $CA = S - I$ 

Выгоды от открытости: 
- Можно сместиться от точки автаркии
- Полезность выше для любого $ r \neq r_A $  
    - Если ставка ниже чем в автаркии, то в 1м периоде импортируем
    - Если ставка выше чем в автаркии, то в 1м периоде экспортируем
  

### ПМОЭ:  реальная ставка

- Для сбережений предполагаем, что  «эффект замещения» доминирует, то есть  $S'(r) > 0$ 
- Для инвестиции при росте ставки труднее занимать капитал, то есть $I'(r) < 0$ 
- $CA'(r) > 0$

Малая экономика:
- Мировая ставка – внешний параметр
- Малая экономика – price taker 
- Изменения в нашей экономике могут сдвигать кривую CA(r), но не влияют на ставку

Большая экономика:
- Изменения в нашей экономике меняют мировую ставку: $CA(r) + CA_{ROW} (r) = 0$


### ПМОЭ: алгебра 

$$ U(C_1,C_2) = \ln(C_1) + \beta \ln(C_2 ) \\ s.t.: C_1 + \frac{C_2}{1+r} = Y_1 + \frac{Y_2}{1+r} $$

$NPV$ потребления = $NPV$ выпуска (эндоумента)

Условия 1-го порядка (уравнение Эйлера):

$$ \frac{\beta  u'(C_2) }{u'(C_2)} = \frac{1}{1+r} $$

В т. максимума потребитель не может увеличить полезность за счет перераспределения потребления между периодами.  Предельная степень заменимости сегодняшнего потребления на завтрашнее потребление = цена единицы потребления завтра в терминах потребления сегодня

- Домашние потребители менее терпеливы, чем остальной мир (ROW): $$ \beta  < \frac{1}{1+r} \Rightarrow C_1>C_2 $$
- Домашние потребители более терпеливы, чем ROW:  $$ \beta  > \frac{1}{1+r} \Rightarrow C_1<C_2 $$
- Домашние потребители также терпеливы, как ROW (полное сглаживание потребления):  $$ \beta  =  \frac{1}{1+r} \Rightarrow C_1=C_2 \\ Y_1 - C_1 = Y_2 - C_2 + rS_1  \\ CA_1 = - CA_2$$

### Глобальные дисбалансы

Глобальные дисбалансы – наличие продолжительных и
значительных профицитов/дефицитов СТО в крупных странах

Др. словами на протяжении длительного времени (десятилетий)
одни страны постоянно чистые кредиторы, а другие – чистые заемщики 

Cumulative  current  account: $$ CCA = \sum_{\tau = t_0}^{t} CA_{\tau}  $$

Чистая международная инвестиционная позиция (Net
International Investment Position): 
 
$$ NIIP = CCA + переоценка\ активов$$ 

## 2.  Валютный курс

- Прямой валютный курс - стоимость единицы нац. валюты в иностранной валюты
- Обратный валютный курс - стоимость единицы иностранной валюты в нац. валюты

Если режим фиксированного (квази-фиксированного) валютного курса:

- Ревальвация нац. валюты - повышение стоимости нац. валюты в терминах иностранной валюты
- Девальвация нац. нац. валюты - снижение стоимости нац. валюты в терминах иностранной валюты


### Структура рынка

Различают три сегмента валютного рынка для валюты каждой страны: 
- Внутренний валютный рынок (on-shore): операции резидентов с резидентам
- Внешний валютный рынок (off-shore): операции нерезидентов с нерезидентам 
- Трансграничные операции с национальной валютой: операции нерезидентов с резидентам


Для участников валютного рынка существует валютный риск: откуда 
потребность в его хеджировании, с помощью форвардных валютных контрактов 

- Спотовый (текущий) валютный курс: курс покупки/продажи валюты немедленно (в тот же день)
- Форвардный валютный курс (F) – заранее оговоренный курс покупки/продажи валюты в определенный момент в будущем (через 1-12 месяцев)

Валютный своп – сочетание двух сделок (спотовая + форвардная или форвардная  + форвардная), обмен одной валюты на другую по оговоренному курсу  с последующим обратным обменом валют по оговоренному форвардному курсу.

Используется для управления ликвидностью (как
операция РЕПО, но где в качестве залога – валюта)

### Реальный валютный курс

- Укрепление национальной валюты (appreciation) - нац. валюта становится дороже, т.е. снижение обратного валютного курса (повышение прямого валютного курса)

- Ослабление национальной валюты (depreciation) - нац. валюта становится дешевле, т.е. повышение обратного валютного курса (снижение прямого валютного курса)


Реальный валютный курс (Real exchange rate) - относительная стоимость корзины иностранных товаров в терминах отечественных товаров: $$RER_t = q_t = \dfrac{S_t P^{*}_t}{P_t}$$


- Укрепление нац. валюты в реальном выражении - увеличение стоимости корзины отечественных товаров относительно стоимости корзины иностранных товаров, $RER_t \downarrow$:

$$ \dot{RER} = \dot{S} - \pi^* - \pi < 0$$

- Ослабление нац. валюты в реальном выражении - уменьшение стоимости корзины отечественных товаров относительно стоимости корзины иностранных товаров 
- увеличивается), $RER_t \uparrow$:

$$ \dot{RER} = \dot{S} - \pi^* - \pi > 0$$

«Внутренний» $ RER = \frac{P_T}{ P_N} $ - мера внутренней
конкурентоспособности торгуемых и неторгуемых товаров
в производстве (конкуренция за ресурсы – факторы
производства) и в потреблении. Укрепление $ RER \Rightarrow $ неторгуемые товары становятся
относительно более дорогими, чем торгуемые


## ППС

Закон одной цены (LOOP) - в долгосрочном периоде цены должны исключать арбитраж в международной торговле, для каждого товара/услуги, торгуемой на мировом рынке, нужно равенство:
$$ P_i = EP^*_i $$

Гипотеза абсолютного паритета покупательной 
способности (PPP) – обменный курс между валютами двух стран должен
уравнивать стоимость репрезентативных корзин, т.е. все уровни цен должны быть одинаковы, если исчислены в одной валюте: 
$$ P = EP^* $$ 

Относительный паритет: $ P = \alpha EP^* $

Объяснения отклонений от паритетов:
- Транспортные и транзакционные издержки, которые в частности могут меняться со временем
- Товары могут быть неоднородными (differentiated products), способность производителей дискриминировать между рынками разных стран 
- Существование неторгуемых товаров (цены не обязаны сравняться, если арбитраж невозможен, теория Балассы-Самуэльссона предсказывает постепенное удорожание реального курса по мере развития экономики страны)
- Монополии и олигополии на рынках товаров
- Технические причины (разные потребительские корзины)
- «Жесткость» цен товаров и услуг

Real effective exchange rate: $$ REER = \prod_{i=1}^N \left( \dfrac {S_i P^{*}_i}{P_i} \right)^{ w_i} $$

Nominal effective exchange: $$ NEER = \prod_{i=1}^N S_i^{ w_i}$$

$w_i$ - вес страны i во внешней торговле (экспорте и
импорте) домашней страны

Выбор ценового индекса:
- ИПЦ (CPI) – цены потребительской корзины, включает также импортные товары
- ИЦП (PPI – цены производителей)  и дефлятор ВВП (цены корзины производства экономики )  включает производимые товары, которые могут мало потребляться внутри экономики 
- индексы цен торгуемых и неторгуемых товаров 


### Долгосрочный (равновесный) RER

Граница производственных возможностей (Production
Possibility Frontier – PPF) – максимально возможная
комбинация производства двух товаров при условии
использования всех ресурсов.

RER задает бюджетное ограничение: $$ GDP=Y_T P_T +Y_N P_N \\ tg(\alpha) = - \frac{P_T}{P_N} $$

Оптимальная комбинация производства $Y_T$ и $Y_N$,
обеспечивающая максимальный ВВП - точка касания PPF и б.о.

Если RER ослабевает (торгуемые товары становятся
относительно более дорогими), то наклон бюджетного
ограничения увеличивается и становится более
выгодно производить больше торгуемых товаров

### Долгосрочный (равновесный) RER = ERER

ERER – equilibrium RER - определяют технологии (PPF), ресурсы (PPF)
и предпочтения потребителей (кривые безразличия).

ERER – RER, согласующийся с
внутренним равновесием ($Y_N = С_N$)
и внешним равновесием ($Y_T = С_T$) 

Эффект Баласса-Самуэльсона (BS)  -  существует положительная зависимость
между уровнями цен и доходами на душу населения в
странах, выраженных в одной и той же валюте

Причины: наличие в экономиках неторгуемых товаров и
разный уровень производительности сектора торгуемых
товаров в богатых и бедных странах





### Модель спроса-предложения нац. товаров

RER определяется из пересечения относительных
спроса и предложения на наши и иностранные товары

$RS = \frac{Y}{Y^*}, RD(\frac{Y}{Y^*},\frac{P^*}{P})$

- Если растет спрос на наши товары: RD вправо,  RER укрепляется
- Если растет наш выпуск: RS вправо, RER слабеет. Интуитивно: наших товаров больше, значит они дешевле

## 3. Модели формирования номинального обменного курса


Отличия номинального курса (S)  от реального (RER): 
- Меньше влияние рынков товаров, больше – рынков финансовых активов (Колебания S влияют на номинальную доходность активов)
- Более короткий временной горизонт (В краткосрочном периоде цены жесткие, так что колебания S определяющие для движений RER)
- Связь с денежно-кредитной и курсовой политикой ЦБ

Долгосрочная перспектива (LR):
- Гибкие (меняются мгновенно) цены
- Плавающий валютный курс
- Главное предположение: раз все гибкое, то выполняется ППС: $ S = \frac{P}{P^*}$

Равновесие на денежных рынках
$$ \frac{M_s}{P}=L(Y, i)$$

- $Y\uparrow \rightarrow L(Y, i)\uparrow $ рост спроса на деньги (кривая спроса на деньги целиком сдвигается вверх)
- $i\uparrow \rightarrow L(Y, i) \downarrow $ снижение спроса на деньги (движение вдоль кривой спроса на деньги)

Номинальный курс определяется относительным
денежным предложением и спросом в 2-х странах: 

$$ S = \frac{ P}{P^*} = \frac{M_s  L(Y^*, i^* )}{L(Y, i) M_s^* } $$

Предположим логарифмическую спецификацию функции спроса на деньги в 2-х странах, тогда: 

$$ m – p = \alpha  + \beta y – \gamma i \\ s = p – p^* = (m–m^* ) – \beta (y–y^* ) + \gamma (i–i^* ) $$ 


где $ m=\ln(M), y=\ln(Y) , p=\ln(P), s=\ln(S) $

Выводы:
- В странах, испытывающих существенный рост денежной массы  будет ослабление нац. валюты
$$ M_s \uparrow ( m\uparrow) \rightarrow P\uparrow \rightarrow S\uparrow \\ M_s^*\uparrow ( m^*\uparrow)\rightarrow P^*\uparrow \rightarrow S\downarrow $$
- Скорость роста цен зависит только от скорости роста предложения денег, а значит скорость изменения курса также зависит только от (относительной) скорости печатания денег
- В странах, испытывающих существенный экономический рост будет укрепление нац. валюты
- В странах, испытывающих существенный рост процентной ставки в LR  будет ослабление нац. валюты: 
$$ i \uparrow \rightarrow L(Y, i)\downarrow \rightarrow P\uparrow [ M_s =const, Y=const, i=\pi^e +r, r=const \Rightarrow  \pi^e \uparrow] \rightarrow S\uparrow \\ 
i^* \uparrow \rightarrow L(Y^*, i^*)\downarrow \rightarrow P^*\uparrow \rightarrow S\downarrow $$
- Влияние ставки i на валютный курс S будет отличаться в SR и LR, т.к.  в LR  изменение происходит  за счет ожидаемой инфляции $\pi^e$, в SR чаще за счет $M^s$



### Подход к курсу через рынок активов : UIP

Инвестиции в два актива – домашний и
иностранный: одинаковая срочность, одинаковая степень рискованности (дефолта), одинаковая ликвидность

$S_t$ - текущий курс рублей за 1 доллар

При условии свободного движения капитала и для
нейтральных к риску (risk-neutral) инвесторов:
- ожидаемая доходность от вложения в два актива, посчитанная с учетом изменения валютного курса, должна быть одинаковой
$$( 1+ i_t ) = (1 + i_t^* ) \frac{ S_{ t+1}^e}{S_t} $$

Если взять логарифмы от обеих частей ($\ln(1+x)\sim x$):
$$i_t = i_t^* + \left( \frac{ S_{ t+1}^e}{S_t}– 1 \right)$$ - это т.н. простая форма UIP  (Непокрытый паритет процентных ставок) 

$\frac{ S_{ t+1}^e}{S_t}– 1$  – ожидаемое ослабление курса рубля по
отношению к доллару

Если $i_t >i_t^* $, то ожидается, что домашняя
валюта будет дешеветь по отношению к иностранной
валюте



### Подход к курсу через рынок активов: CIP

CIP (покрытый паритет процентных ставок):

- Будущий рыночный курс $ S_t+1 $ неизвестен, и инвестор несет валютный риск 
- Чтобы его избежать, инвестор может подписать форвардный контракт по определенной цене $F_t$
- Тогда при отсутствии ограничений на движение капитала и при отсутствии арбитража выполняется равенство безрисковых доходностей двух активов:
$$(1+ i_t )= (1+ i_t^* )\frac{ F_t}{ S_t} \\ i_t = i_t^* + \left( \frac{ F_{ t}}{S_t}– 1 \right)$$

Если $ \frac{F_{ t}}{S_t}– 1 > 0 $, то говорят, что домашняя валюта
продается с форвардным дискаунтом (т.е. форвардный
курс домашней валюты слабее, чем спотовый курс) или
наоборот, иностранная валюта продается с форвардной
премией

CIP может не выполняться если:
- существуют ограничения на движение капитала (или угроза ввода)
- разные, по мнению рынка, риски дефолта национальных контрагентов

Премия на ограничение движения капитала или
страновая премия:
$$ (1+ i_t - СRP_t )= (1+ i_t^* ) \frac{F_t}{S_t} \\  CRP_t = i_t - i_t^* - \left(\frac{F_t}{S_t} -1\right) $$


- $ CRP_t >0$ ограничения на движение капитала не дают некоторым инвесторам зайти в домашние активы, доходность домашних активов могла бы стать ниже с приходом новых инвестров и притоком капитала
- $ CRP_t <0$ ограничения на движение капитала не дают держателям домашних активов легко «избавиться» от них, доходность домашних активов стала бы выше, если бы часть держателей смогла бы выйти 


Если UIP & CIP выполняются, то форвардный курс должен равняться ожидаемому курсу с учетом срочности:
$$F_{t,t+k} = E_t (S_{t+k} ) $$




### Подход к курсу через рынок активов: RP

UIP должно быть модифицировано таким образом, чтобы учитывать валютный риск: чем более волатилен валютный курс, тем больше должна быть компенсация за риск 

- Для домашнего инвестора, покупающего иностранные бумаги, доходность на иностранные бумаги должна превосходить доходность на домашние бумаги
(рассчитанную в домашней валюте) на премию за риск (RP t >0):
$$ (1+ i_t )= (1+ i_t^* - RP_t ) \frac{S_{t+1}^e}{S_t} $$ 

- Для иностранного инвестора:  инвестирование в наши домашние бумаги должно превосходить доходность инвестирования в свои активы: 
$$ (1+ i_t -  RP_t^* )= (1+ i_t^*)\frac{S_{t+1}^e}{S_t} $$ 


Можем определить премию за риск $RP_t$ как отклонение от UIP: 
$$ RP_t = \left(\frac{S_{t+1}^e}{S_t} -1\right) - ( i_t - i_t^* ) $$

Если CIP выполняется (при отсутствии ограничений
на движение капитала), то
$$ i_t - i_t^* =\left(\frac{F_t}{S_t} -1\right) \\  RP_t = \left(\frac{S_{t+1}^e}{S_t} -1\right) - \left(\frac{F_t}{S_t} -1\right) = \frac{S_{t+1}^e - F_t}{S_t}$$
 

$RP_t$   - форвардное смещение (forward bias), отделяет ожидаемый курс валюты от форвардного курса. 

Ошибка прогноза спотового курса $\frac{S_{t+1} - S_{t+1}^e}{S_t} $ .

$$ \frac{S_{t+1} - F_{t}}{S_t} = \frac{S_{t+1}^e - F_t}{S_t} +   \frac{S_{t+1} - S_{t+1}^e}{S_t} $$ 

Форвардный курс может плохо предсказывать будущий спотовый курс из-за:
- существования премии за риск 
- ошибок рынка в прогнозе будущего спотового курса.



### Равновесие на валютном рынке в SR и LR

$$( 1+ i_t ) = (1 + i_t^* ) \frac{ S_{ t+1}^e}{S_t} \\ S_t = UIP(\cdot) = S_{t+1}^e \frac{1+ i_t^*}{1+ i_t} $$



Условие равновесия на валютном рынке (UIP):
$$ i_t = UIP(\cdot)  $$

где $i_t$ -  доходность
домашнего актива,  $i_t = i_t^*  +  \frac{S_{t+1}^e}{S_{t+1} }-1 $ - ожидаемая доходность иностранного актива (в домашней валюте)

- $i_t \downarrow  \rightarrow  S_t \uparrow $: cнижаются национальные ставки $\Rightarrow$ национальный актив менее доходен $\Rightarrow$ инвесторы уходят в другие активы, то есть в другую валюту $\Rightarrow$ национальная валюта дешевеет




Валютный курс - вперед-смотрящая переменная: 
$$ S_t = \frac{1+i^*_t}{1+i_t} \dots \frac{1+i^{e*}_{t+n}}{1+i^{e}_{t+n}} E_t(S_{t+n+1}) $$

- $S_t$ реагирует на все новости и слухи о будущем
- Волатильная переменная, которую сложно предсказать
- На ежедневной основе можно представить $S_t$ как случайное блуждание (random walk): 
$$S_t = S_{t-1 } + e_t, где\ e_t \sim N(0, \sigma^2) \\  E_t [S_{t+k} ] = E_t [S_t + e_{t + 1} + \dots  + e_{t+k} ] = S_t $$







### Валютный курс через фундаментальные факторы: 

Номинальный курс = дисконтированная сумма ожиданий
функции от фундаментальных факторов

Из монетарного подхода:

$$ s_t = p_t – p_t^* = (m_t–m_t^* ) – \beta (y_t–y_t^* ) + \gamma (i_t–i^*_t ) $$ 


Из UIP:  
$$ i_t -  i_t^*  =   s_{ t+1}^e - s_t $$

Подставим в монетарный подход и получим: 

$$ s_t =\frac{(m_t–m_t^* ) – \beta (y_t–y_t^* ) + \gamma s_{ t+1}^e }{1 + \gamma} = Z_t + \alpha s_{ t+1}^e = \dots $$

$Z_t$  содержит в себе т.н. фундаментальные факторы (динамику денежной массы и экономический рост)

### Carry trade

Сarry trade - инвестиционная стратегия международных
инвесторов 
- занять в валюте с низкой $i_t $ 
- купить актив в валюте с высокой $i_t^* $ 


Если $s_{t+k}  - s_t = \alpha + \beta (i_{t,k} – i_{t,k}^* )+e_{t,t+k}$
и $\beta<1$,  то имеет
смысл инвестировать в валюту с более высокой
доходностью 

Особенно выгодна стратегия carry trade, когда
высокодоходная валюта продолжает укрепляться, при этом приток капитала в carry trade валюту
способствует ее укреплению

Периоды, когда сarry trade приносит большую прибыль,
сменяется периодами больших убытков, когда
высокодоходная валюта начинает ослабевать или
начинает «лопаться пузырь», например, рынка
недвижимости – “unwinding of the carry trade”.


### Forward premium puzzle

Парадокс форвардной премии: $F_t$ – смещенная оценка $S_{t+1}$

Объяснения:
1. Не удается смоделировать столь волатильную премию за риск, какая наблюдается эмпирически
2. Участники рынка – не вполне рациональные агенты (noise trader)
3.  $S_{t+1}^e \neq  E(S_{t+1 })$,  например, если распределения шоков оценивается инвесторами отличным от экономистов образом
4. Peso-problem: сравнительно небольшая вероятность негативного сценария (резкого ослабления кэрри-трэйд валюты), которое не происходит в период доступной временной выборки
5. Simultaneity bias


### Денежный и валютный рынки в SR

На денежном рынке:  $ \frac{M^s}{P} = L(Y, i)$ 

На валютном рынке:   $( 1+ i_t ) = (1 + i_t^* ) \frac{ S_{ t+1}^e}{S_t} $ - доходность домашнено актива = ожидаемая доходность зарубежного актива

- $ M^s \uparrow $ (временно) $\rightarrow [M/P] \uparrow (P =const)  $ избыток предложения реального денежного баланса $\rightarrow i \downarrow (Y =const) \rightarrow  $ не выполняется UIP $ \rightarrow  $ снижение спроса на домашние и рост спроса на зарубежные активы, снижение спроса на рубль и рост спроса на доллар  $ \rightarrow   S\uparrow  [S_{t+1}^e =const]$  рубль слабеет к доллару, так что выполняется UIP: $ i= i^*+ [S_{t+1}^e / S -1] $

В долгосрочной перспективе (LR):
- $Y_{LR} = Y_p$ - устанавливается потенциальный (естественный) уровень выпуска,  естественный уровень безработицы
- Цены и зарплаты – гибкие, мгновенно меняются для поддержания равновесия на рынке товаров, услуг и труда




### Модель «перелета» (overshooting) курса Дорнбуша


Необратимое изменение денежной массы в LR:
в SR валютный курс «перелетает» свое LR
значение за счет жесткости цен в SR


- SR: $ M^s \uparrow $ (permanent) $\rightarrow \dots \rightarrow  S\uparrow $
- SR:  $ M^s \uparrow $ (permanent)$\rightarrow P^e_{LR} \uparrow $ ППС в LR $\rightarrow S_{LR}^e \uparrow$ сдвиг UIP $\rightarrow S\uparrow \uparrow  (больше, чем при временном M^s \uparrow)$
- LR:  $ M^s \uparrow $ (permanent) $P \uparrow [M/P]\downarrow (M =const) $ избыточный спроса на реальный денежный баланс $ \rightarrow i \uparrow \ rightarrow$ спрос на домашние активы $uparrow$, спрос на рубль $ \uparrow \rightarrow S\downarrow$, рубль укрепляется при переходе в LR
- $i$ возвращается к изначальному значению
