Чистый оборотный капитал (NWC):

$$NWC = Текущие \ активы – Текущие \ обязательства $$

Profitability index: 

$$PI = \frac{PV \ of \ future \ CF}{Initial \ Inv} = \frac{1}{I} \sum{\frac{CF_t}{(1+i)^t}}$$

Несложно заметить, что при оценке проектов, предусматривающих одинаковый объем первоначальных инвестиций, критерий PI полностью согласован с критерием NPV.

Точный дисконтированный период окупаемости (precise discounted payback period, PDPBP) – минимальное число лет (периодов), в течение которых прогнозируемые дисконтированные денежные потоки в общей сложности устойчиво достигают величины, не меньшей величины первоначальных вложений.

$$PDPBP = t - 1 + \frac{|CDCF_{t-1} |}{DCF_t}$$

$CDCF_{t-1}$ – накопленная сумма дисконтированных денежных потоков к моменту времени $t-1$, $DCF_t$  – дисконтированный денежный поток 


Net realizable value (NRV) is the value of an asset that can be realized upon the sale of the asset, less a reasonable estimate of the costs associated with either the eventual sale or the disposal of the asset in question.


###  WACC

Средневзвешенные затраты на капитал (WACC) Weighted average cost of capital
-  та предельная рентабельность деятельности компании, которая должна удовлетворять требованиям всех заинтересованных сторон
- отражает стоимость вновь привлекаемого объема финансирования, поэтому эта величина может отличаться от стоимости уже привлеченного капитала

$$WACC = w_d r_d (1-t) + w_{pref} r_{pref} + w_e r_e$$

$$WACC =  \frac{D}{K} r_d (1-t) +  \frac{PS}{K} r_{pref} + \frac{E}{K} r_e$$

$$K  = E+D+PS$$ 

$w_d$ - доля заимствованных средств (заемный кап. к собственному кап.), 
$w_{pref}$ - привелигированные акции в общей доле финансирования,
$w_e$ - доля собственного капитала,
$t$ - ставка налога, 
$r_{pref} = \frac{D_{pref}}{P_{pref}(1-F)}$, $D_{pref}$ - дивиденды $P_{pref}$- рыночная цена акций,  $F$ - фиксированные затраты на организацию эмиссии в процентах  

Принципы расчета:
- ставка дисконтирования корректна, если структура капитала относительно стабильна
- используются текущие или ожиданиемые стоимости привлечения капитала
- в качестве ставки дисконтирования не вычитать процент, но учитыватьь налог
- расчитывается в номинальном вражении
- в той же валюте, чтои потоки

Факторы:
- неподконтрольные (уровень ставок на рынке/альтернативная доходность, риск премия, налоговые ставки)
- подконтрольные (целевая структура финансирования, дивидендная политика, инвестиционная политика)

Задолженности в расчете WACC:
- типичные долгосрочные проценты: кредиты, облигацигации
- краткосрочные заимствования (если используются на постоянной основе)
- лизинговые платежи (если долгосрочный лизинговый контракт)

Не включаются в расчет:
- краткосрочные заимствования временного характера
- беспроцентные источники финансирования 

$r_d$ - банковская ставка (на практике постоянна)




$$v = \sum \frac{C*mV}{(1+YTM)^t} + \frac{mV}{(1+YTM)^T}$$

$v$ - рыночная стоимость облигации, $mV$ - maturity value


Упрощенная формула:
$$YTM = \frac{c+\frac{mv-p}{T}}{\frac{mv+p}{2}}$$
$p$ - рыночная стоимость

Модификация:
$$YTM = \frac{c+\frac{mv-p}{T}}{\frac{mv+p}{3}}$$


Аннуитет: $$PVA = A + \frac{R}{P} \frac{1 - (1+\frac{r}{m})^{mn}}{(1+\frac{r}{m})^{m/n}-1}$$

$PVA$ - приведенная стоимость всех приведенных платажей (стоимость лизингового имущества), $A$ - аванс (если есть), $R$ - годовой лизинговый платеж, $P$ - кол-во в течение года, $n$ - кол-во лет, $m$ - кол-во начислений

### Модель дисконтирования дивидендов


$$P = \sum_{t=1}^{\infty} \frac{D_0 (1+g)^t}{(1+r_e)^t}$$

Модель Гордона в жизни маловероятна (на практике $g$  нестабилен)

$$P_0 = D_0 \frac{1+g}{r_e - g}$$

Собственный капитал: 

$$r_i = D + \frac{1+g}{P} + g$$

CAPM:

$$\beta_l = \beta_u (1+(1-t)\frac{D}{E}) $$

$\beta_l$ - leveraged, $\beta_u$ - unleveraged

$$r_d<r_{pref}<r_e$$

На основе WACC строится график предельных затрат на капитал
(затрат на привлечение последней единицы доп. ср-в финансирования)



- по x - необходимый объем привлекаемых средств 
- по y - стоимость привлечения (доходность для инвесторов)

Чем больше источников финансирования, тем дороже привлечение нового капитала


### Структура капитала

Схема расчета риска по денежному потоку:
-  валовая (маржинальная) прибыль = выручка - переменные издержки =  $TR - VC = \sum p_i  Q_i - \sum vc_i  Q_i$
- постоянные издержки: $FC$ 
- выплата процентов: $I$ 
- corporate tax: $T_c$
- net operating profit after taxes (чистая прибыль после уплаты налогов): $NOPAT$
- дивиденды по привелегированнным акциям: $DIV_{pref}$
- дивиденды по обыкновенным акциям: $DIV_{com}$
- нераспределенная прибыль: $NI$


$$EBIT = TR - FC - VC $$

$$NOPAT = EBIT - I - T_c $$

$$NI = NOPAT - DIV_{pref} - DIV_{com}$$

Рентабельность:
$$ROE = \frac{NI}{F} $$ 

Операционный риск зависит от внешних факторов:
- конъюнктура рынка
- стратегии конкурентов
- структура спроса и внутренние факторы
- соотношение переменных и постоянных издержек

Операционный рычаг - показатель, позволяющий измерить степень влияния операционных расходов на прибыль компании, чувствительность операционной прибыли $EBIT$ к $\Delta$-выручки (в объеме продаж)

$$DOL = \frac{\Delta \% EBIT}{\Delta \% S}$$

$\Delta \% EBIT$ - процентное изменение прибыли, $\Delta \% S$ - процентное изменение выручки

$$DOL = \frac{Q(P-V)}{Q(P-V) - F}$$

$V$ - variable cost, $F$ - fixed cost




$D/E$ - соотношение долгового капитала к собственному

Финансовый рычаг - показатель позволяет измерить степень влияния фиксированных платежей на чистую прибыль компании

$$DFL = \frac{\Delta \% EPS}{\Delta \% EBIT}$$

$EPS$ - earnings per share

$$DFL = \frac{EBIT}{EBIT - I - \frac{PD}{1-t}}$$

$PD$ - дивиденды по привелигированным акциям


Совокупный рычаг

$$DTL = DOL * DFL = \frac{\Delta \% EPS}{\Delta \% S} = \frac{Q(P-V)}{Q(P-V) - F - I - \frac{PD}{1-t}} $$


### Анализ безубыточности деятельности компании

Точка безубыточности - уровень продаж, когда выручка равна затратам (инвестору безразлично реализовывать проект или нет)


Объем продаж  в точке безубыточности:

$$EBIT = Q(P-V) - F = 0 $$

$$Q_{be} = \frac{F}{P-V}$$

Если постоянные издержки находятся на низком уровне, подушка безопасности выше

Точка безразличия - такое значение EBIT, при котором чистая прибыль на вложенный собственный капитал EPI будет одинакоыв независимой от всех вариантов финансирования (такое состояние здолженности компании, при котором  издержки, связанные с использованием доп. фактора производства устанавливаются выручкой их прироста продукции)

$$EPS =  \frac{(EBIT-I)(1-t)-PD}{Sh_{com}}$$

### Дивидендная политика

Дивиденд - любой доход, направляемый акционерам от организации, при распределении прибыли, остающейся после налогооблажения, выплат  процентов по привелигированным акциям, по приндлежащим акционеру акциям пропорционально акционеров

Внешние факторы:
- уровень налогообложения доходов  в форме дивидендов и в форме прироста стоимости капитала компании
- законодательные требования
- ограничения со стороны держаетелей целевых обязательств 
- затраты на внешнее финансирование
- асимметрия информации

Внутренние факторы:
- структура капитла компании
- нехватка ликвидности средств
- нестабильность чистой прибыли компании
- инвсетиционная политика компании
- стимулы развития бизнеса, принадлежность к отрасли  


### Теория предпочтительности дивидендов (Гордон)

- инвесторы предпочитают опр. дивиденд вместо неопр. прироста цены акции в будущем
- каждая ед. текущего дохода, выплачиваемая в форме дивидендов, очищена от рисков, а значит, стоит всегда больше, чем доходЮ отложенный на будущее

$r_{d}$ - ставка доходности заемного капитала,  $(r_{eu}-v_d) \frac{D}{E} (1 - t_c)$ - риск премия

3) Налоги и доли (учитывая подоходный налог)

$$CF_{инвесторы} = (EBIT - I)(1-T_c)(1 - T_e) + I(1 - T_d) $$


$(EBIT - I)(1-T_c)(1 - T_e)$ - часть инв. денежного потока для акцинеров

- Если ставка налога на доход по деловым обязательствам равна ставке налога по долевым обязательствам,  то получаем результаты теоремы 2
- Если $T_e<T_d$, выигр от заемного финансирования сохраняется (но выигрыш меньше на $T_e D$)

$$V_L=V_u+TD$$

- Если $(1-T_c)(1 - T_e) =(1 - T_d)$, то выгоды от использования собственного капитала и заемного уравниваются и инвестор несет одинаковые налоговые выплаты

### Модель целевого остатка ден. ср-в:

Предпосылки:
- все требования к оплате исполняются немедленно
- в каждом периоде спрос на деньги стабилен
- остатки ден. ср-в используются равномерно
- в конце каждого периода сальдо ден. ср-в обнуляется
- транзакц. издержки постоянны

Идея: оптимальная сумма денежных средств определяется исходя  из  цели мин. транзакционных и альтернтивных издержек

 $$TC = F \frac{T}{C} + r \frac{C}{2}$$
 
$F$ - пост. транз. изд.,  $T$ - потребность за период в ден. ср., $C$ - объем пополняемых компанией ден. ср-в за счет продажи ц.б. или привлечения займа,  $r$ - стоимость альт. изд. хранения денег


$$ C^*= \sqrt{\frac{2TF}{r}} $$

$C^*$ - оптимальная величина ден. ср-в,  которые необходимо иметь на счете компании

###  Модель Миллера целевого остатка ден. ср-в

- 2 актива: ден. ср-ва и ликвидные ц.б.
- транз. изд. постоянны, зависят от объема сделки
- остатки ден. ср-в суммируются случ. образом

Идея: в период, когда остатки д.с. достигают $H$, она покупает ликв. ц.б.  на сумму  $H-Z$,  чтобы вернуть остаток д. с. к целевому уровню (или,  наоборот, покупает $L-Z$)

Целевой остаток:
$$Z  = (\frac{3F\sigma^2}{4r})^{1/3} + L$$

Верхняя граница:

$$H = 3 (\frac{3F\sigma^2}{4r})^{1/3} + L = 3Z - 2L $$
