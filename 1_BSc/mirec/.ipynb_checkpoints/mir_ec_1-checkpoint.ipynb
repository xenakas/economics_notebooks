## 1. Система национальных счетов и платежный баланс

### Открытость экономики: товарные рынки

Экспорт товаров и услуг – Импорт товаров и услуг = Баланс товаров и услуг:

$$X–M=TB$$


Степень открытости товарного рынка в стране – экспорт и импорт товаров (и услуг), в % ВВП 

Малые открытые экономики – большая доля экспорта и импорта в ВВП, доля в мировом ВВП не большая

Закрытые экономики - низкая доля торговли в ВВП (США близка к таковым из-за удаленности и размера)

Экспорт и импорт могут быть больше, чем ВВП страны (пример – Сингапур), т.к. экспорт и импорт могут включать промежуточные товары, а ВВП – конечные товары, небольшие по размеру удачно расположенные (и с офшорным статусом) страны активно занимаются ре- экспортом

Торгуемые товары (экспортируемые, импортируемые) - конкурируют с иностранными товарами на внешних и внутренних рынках

При открытости товарных рынков выбор между отечественными и иностранными товарами влияет на выпуск в экономике:

спрос на отечественные товары↑ → выпуск (ВВП)↑

Главную роль в выборе между отечественными и иностранными товарами играют относительные цены на эти товары, посчитанные в единой валюте – реальный валютный курс

Открытость финансовых рынков позволяет:

- финансировать дефициты торговых балансов (точнее СТО) за счет притока иностранного капитала в страну, представляющего собой заимствования у иностранных экономических агентов

- сберегать в виде иностранных активов/давать в долг иностранным экономическим агентам образующиеся профициты торговых балансов (СТО) 

### Национальные счета

ВВП/GDP - стоимость всех конечных товаров и услуг, произведенных в экономике за определенный период времени (т.е. географический принцип):

$$ВВП=C+I+G+X−M$$


ВНП/GNP - стоимость всех конечных товаров и услуг, произведенных принадлежащими отечественным эконом. агентам (резидентам) факторами производства (трудом, капиталом) за определенный период времени (т.е. принцип собственности):

$$ВНП=ВВП+NFP$$


NFP - net factor payments from abroad


У страны, являющейся чистым должником остального мира, ВНП, вероятнее всего,  не превосходит ВВП: 
чистый должник платит проценты по долгу. Это факторные платежи, которые входят с минусом в счет текущих операций, а значит уменьшают ВНП по сравнению с ВВП. Другими словами, нация не получает в доход часть того, что произвела, потому что приходится платить проценты.

Доходы факторов производства за рубежом: 

- Зарплаты резидентов, работающих в др. странах
- Прибыль компаний от производства в др. странах


Cчет текущих операций (CA) -  это чистый экспорт плюс чистые доходы на факторы

$$CA = X-M+NFP +NT $$

NT - баланс текущих трансфертов (net unilateral current transfers)

Если $NT=0$, то  $ВНП =C+I+G+CA$ 


### Платежный баланс

Платежный баланс (Balance of Payments, BoP) – отражает транзакции страны с остальным миром:

- Покупка-продажа товаров и услуг
- Покупка-продажа финансовых активов

Платежный баланс имеет 2 главных счета:
- Счет текущих операций (current account, CA)
- Финансовый счет (KA)

Как правило, транзакция с товаром (услугой) зеркально
отражается в транзакциях с финансовыми активами. В теории: $$CA=KA$$


В реальности:
- $CA<0$ хорошо, если покрывает временный спад выпуска или покрывает рост спроса на инвестиции, долги потом отдадим
- $CA<0$ плохо, если только поддерживает потребление, ведь не создается базы для возвращения долгов
- $CA>0$ хорошо, если временный рост выпуска – формируем резерв на будущее
- $CA>0$ плохо, если домашние инвестиции при этом падают, это сигнал о том что возможностей для внутренних инвестиций мало и резиденты предпочитают инвестировать в другие страны


_Страны с чистым притоком капитала должны иметь дефицит счета текущих операций:_ так как СТО и финансовый счет это две основных статьи платежного баланса, то в целом дефицит СТО должен финансироваться прежде всего притоком капитала. Но в конкретных случаях может быть и не так, если например меняются резервы ЦБ (финансируем дефицит СТО из резервов), или происходят транзакции по счету капитала (получаем гуманитарную помощь). Утверждение скорее верное в долгосрочном плане, но неверное технически. 

_США практически ничего не стоит сколь угодно долго финансировать дефицит счета текущих операций страны в силу использования доллара США в качестве мировой резервной валюты_ : Нужны 2 условия. а) пока в мировой экономике есть спрос на доллар как средство сбережения и альтернативную валюту и b) пока мировая экономика растет, то всегда будет генерироваться дополнительный спрос на новые доллары. И в такой ситуации США ничего не стоит удовлетворять этот спрос напечатанными долларами (которые суть чистые иностранные активы для других стран) 



На практике:  




$$CA  +  CO + E\& O = KA
$$

Счет операций с капиталом (СО) это приобретение/выбытие непроизведенных финансовых активов и капитальные трансферы (в частности прощение долгов).

Счет операций с капиталом:
- Продажа – покупка непроизведенных нефинансовых активов (патентов, лицензий)
- Чистое прощение иностранного долга (государственным и частным секторами)
- Гуманитарная помощь, безвозмездная передача и т.п.


Чистые ошибки и пропуски (E & O ) – по остаточному принципу, чтобы все компоненты платежного баланса сложились в ноль: не удается
правильно учесть все транзакции в статистике (часто ошибки
или намеренные искажения в статистике импорта, в оттоке
капитала, но ЦБ учатся их корректировать)


$$CA + CO = KA + \Delta RES$$ 

__Rule of thumb__: Если у вас возникают проблемы с тем, чтобы понять знак какой-то транзакции в ПБ, представьте какая торговая транзакция (экспорт/импорт) ее зеркалит в балансе. Например, гражданин купил иностранные акции, чтобы купить акции, надо чем-то заплатить иностранцам – для этого подходит экспорт, который дает плюс в СА. Значит наша транзакция - это зеркальный плюс в КА. А вторая часть транзакции зависит от того, чем именно он за эти акции заплатил.


### Финансовый счет (движения капитала)


Финансовый счет без резервов (KA, он же KA без резервов) это изменение иностранных активов в собственности наших граждан минус изменение иностранных обязательств наших граждан.


$$ KA  = \Delta NFA = + \Delta (иностранные\ активы)- \Delta (иностранные \ обязательства)
$$

$NFA $ - Net foreign assets или изменение чистых
иностранных активов


$KA<0$ - чистый приток капитала, приток в страну иностранного капитала больше, чем отток из страны отечественного капитала заграницу $\rightarrow$ чистый внешний долг$\downarrow$

$KA>0$ - чистый отток капитала,  приток в страну иностранного капитала меньше, чем отток из страны отечественного капитала заграницу  $\rightarrow$ чистый внешний долг$\uparrow$

_У экономики, в которой в текущий момент времени наблюдается положительная величина NFA_:
Предположим, что переоценка активов отсутствует. Иначе можно представить ситуацию, когда активы внезапно обесценятся, и тогда их уже не получится в какой-то момент использовать, чтобы купить побольше импорта.
Плюс чистых активов означает, что в какой-то момент их можно обменять на иностранные товары и услуги, таким образом ухудшив баланс счета текущих операций  
Другими словами, так как $CA=\Delta NFA$, а в долгосрочном равновесии нет смысла иметь $NFA>0$, то будем момент, когда активы будут уменьшаться. Иначе зачем мы их копили.



### Платежный баланс для страны нерезервной валюты

$$CA = KA^{к.р.} + \Delta RES$$

$KA$  кроме резервов -  это чистый приток-отток капитала по каналам частного сектора и госсектора, за
исключением операций ЦБ с международными резервами
(резервными активами)

$\Delta RES$ – изменение международных (валютных) резервов (резервных
активов)

- $CA-KA^{к.р.} < 0$ значит, что ЦБ за счет снижения международных резервов компенсирует дефицит валюты, поступающей в экономику по каналам торговли товарами и услугами (СA) и финансовыми активами (KA)
- $CA-KA^{к.р.} > 0$  значит, что ЦБ за счет увеличения международных резервов компенсирует избыток валюты, поступающей в экономику по каналам торговли товарами и услугами и финансовыми активами

### Платежный баланс для страны резервной валюты

$$ CA=KA^{к.р.} + \Delta RES_{Наш \ ЦБ}  - \Delta RES_{ ЦБ \ др.стран \ в\  валюте\  этой\  страны}$$

Как правило, у страны резервной валюты $ \Delta RES \approx 0 $

ЦБ др. стран покупают резервную валюту этой страны, что может служить основным или дополнительным
источников финансирования дефицита счета текущих
операций (кроме чистого притока инвестиций в
нерезервные финансовые активы, $KA^{к.р.} < 0$)

###  Национальные сбережения

Национальные сбережения - то что мы произвели, но не потребили 
$$S = Y-C-G = I + CA $$
в закрытой экономике $CA=0$

$S_p = Y-T-C =  I_p - BB + CA$ -  сбережения частного сектора

- На финансирование частных инвестиций
- На финансирование дефицита гос. Бюджета
- На приобретение иностранных активов


$S_g = T – G$ -  сбережения государства

$BB = T – G – I_g$ -  баланс госбюджета (budget balance)




###  Intertemporal approach to CA

Межвременная тоговля – обмен ресурсами во времени

### Простая модель открытой экономики (ПМОЭ)

В открытой экономике можно давать или брать в долг у остального мира по
заданной извне процентной ставке $r$

$$ C_1 + S_1 = Y_1 \\  C_2 = Y_2 + S_2 = Y_2 +  (1+r)S_1  = Y_2 + (1+r)(Y_1 -C_1 )$$ 

Откуда  получаем  уравнение прямой бюджетного ограничения  для задачи потребителя:

$$ \max U(C_1,C_2 ) \\ s.t.: (1+r)C_1 + С_2 = (1+r)Y_1 + Y_2 $$





### ПМОЭ: счет текущих операций

Сбережения $S_1$ это счет текущих операций $CA_1$
- инвестиций в 1м периоде нет, $CA = S - I$ 

Выгоды от открытости: 
- Можно сместиться от точки автаркии
- Полезность выше для любого $ r \neq r_A $  
    - Если ставка ниже чем в автаркии, то в 1м периоде импортируем
    - Если ставка выше чем в автаркии, то в 1м периоде экспортируем
  

### ПМОЭ:  реальная ставка

- Для сбережений предполагаем, что  «эффект замещения» доминирует, то есть  $S'(r) > 0$ 
- Для инвестиции при росте ставки труднее занимать капитал, то есть $I'(r) < 0$ 
- $CA'(r) > 0$

Малая экономика:
- Мировая ставка – внешний параметр
- Малая экономика – price taker 
- Изменения в нашей экономике могут сдвигать кривую CA(r), но не влияют на ставку

Большая экономика:
- Изменения в нашей экономике меняют мировую ставку: $CA(r) + CA_{ROW} (r) = 0$


### ПМОЭ: алгебра 

$$ U(C_1,C_2) = \ln(C_1) + \beta \ln(C_2 ) \\ s.t.: C_1 + \frac{C_2}{1+r} = Y_1 + \frac{Y_2}{1+r} $$

$NPV$ потребления = $NPV$ выпуска (эндоумента)

Условия 1-го порядка (уравнение Эйлера):

$$ \frac{\beta  u'(C_2) }{u'(C_1)} = \frac{1}{1+r} $$

В т. максимума потребитель не может увеличить полезность за счет перераспределения потребления между периодами.  Предельная степень заменимости сегодняшнего потребления на завтрашнее потребление = цена единицы потребления завтра в терминах потребления сегодня

- Домашние потребители менее терпеливы, чем остальной мир (ROW): $$ \beta  < \frac{1}{1+r} \Rightarrow C_1>C_2 $$
- Домашние потребители более терпеливы, чем ROW:  $$ \beta  > \frac{1}{1+r} \Rightarrow C_1<C_2 $$
- Домашние потребители также терпеливы, как ROW (полное сглаживание потребления):  $$ \beta  =  \frac{1}{1+r} \Rightarrow C_1=C_2 \\ Y_1 - C_1 = Y_2 - C_2 + rS_1  \\ CA_1 = - CA_2$$

### Глобальные дисбалансы

Глобальные дисбалансы – наличие продолжительных и
значительных профицитов/дефицитов СТО в крупных странах

Др. словами на протяжении длительного времени (десятилетий)
одни страны постоянно чистые кредиторы, а другие – чистые заемщики 

Cumulative  current  account: $$ CCA = \sum_{\tau = t_0}^{t} CA_{\tau}  $$

Чистая международная инвестиционная позиция (Net
International Investment Position): 
 
$$ NIIP = CCA + переоценка\ активов$$ 

##### Задача:

Налоги забираются из эндаумента перед тем как потребитель делает решение о потреблении. Соответственно потребитель выбирает уровень сбережений в первом периоде в виде:
$$S1 = Y1-T1-C1$$
$$C2 = Y2-T2 + (1+r)S1$$

БО:  $Y2-T2-С2+(1+r)(Y1-T1-C1) = 0 $

Так как потребитель никак не влияет на $Ti$ и $Gi$, то он воспринимает их как константы, и по сути задача потребителя это максимизация $u(C1)+ \beta u(C2)$ при эндаументе $Y1-T1, Y2-T2$, то есть классическая форма модели межвременного выбора. 

Заметим, что тут предполагается, что потребитель ведет себя как ведомый по Штакельбергу. Это резонное предположение для групп людей («от меня ничего не зависит»).


На графике в пространстве $С1,С2$ линия бюджетного ограничения будет проходить через точку $Y1-T1, Y2-T2$ с наклоном -$(1+r)$ – множителем при $С1$

Решение задачи потребителя в общем виде:

Лагранжиан задачи максимизации: $L = u(C1)+ \beta u(C2) + \lambda (Y2-T2-С2+(1+r)(Y1-T1-C1))$ 

Условия первого порядка: $\partial L/ \partial  Ci = 0$, то есть $u’(C1)-\lambda(1+r) = 0$ и $\beta u’(C2)- \lambda =0$ 

Исключая $\lambda$, получаем уравнение Эйлера $u’(C1)/u’(C2)= \beta (1+r) $

Если $\beta=1/(1+r)$ то $u’(C1)=u’(C2)$ и значит $С1=С2$, потребление полностью сглаживается

__СТО в 1-м (CA1) и 2-м (CA2) периодах__ 


$$CA = S-I$$

Инвестиции нулевые по условию. Значит $САi = Si$

Сначала вычислим сбережения (доход минус съеденное) в каждом из периодов: 

$$S = ВНП–C–G$$

Заметим, что $ВНПi$ далеко не всегда совпадает с $Yi$, так как $ ВНП=ВВП + NFP $  и во 2м периоде это важно.


Сбережения в 1м периоде: $$S1 = ВНП-C1-G1 = Y1-T1-C1$$
Сбережения во 2м периоде: $$S2 = ВНП-С2-G2 = Y2-T2-C2 + rS1$$

Подставляем $Y2-T2-C2$ из бюджетного ограничения и получаем

$$S2 = -(1+r)(Y1-T1-C1) +rS1 = -(1+r)S1 +rS1 = -S1 $$

В результате $CA2 = -CA1$






__Если экономика изолирована от остального мира: __

В автаркии $Ci = Yi-Ti$, так как больше товары брать неоткуда, и некуда отдавать. 

Соответственно подставляем $Ci$ в уравнение Эйлера и получаем 

$$r = u’(Y1-G1)/\beta u’(Y2-G2) -1$$

Заметим, что реальная ставка ничем не ограничена, может быть как положительной, так и отрицательной. Отрицательную ставку интуитивно можно воспринимать как плату за перенос хоть какой-то части товара во 2й период. Если много желающих перенести, то этот фактор перевесит желание агентов  пользоваться товарами в 1м периоде, которое повышает рыночную ставку.


__Пусть $\beta =1/(1+r)$  и изначально ожидалось, что $Y1-G1=Y2-G2$, но затем происходит рост госрасходов в 1-м периоде на $\Delta G$__ 


Если  $\beta =1/(1+r)$,  С1 = С2 при любых движениях эндаумента. 

Изначально $Y1-G1 = Y2-G2 = С1 = С2 $, что значит нулевые сбережения в 1м периоде и $CAi = 0$

Если G1 растет, это уменьшает располагаемый доход частного сектора, а значит, будет снижаться и уровень потребления в обоих периодах $C1=C2$. 
Значит, во 2м периоде становится $Y2-G2>C2$ и соответственно $CA2>0$. Отсюда следует, что $CA1<0$ и $С1$ падает меньше, чем растет $G1$.



__Пусть $\beta =1/(1+r)$  и изначально ожидалось, что $Y1-G1=Y2-G2$, но затем происходит рост госрасходов в 2-м периоде на $\Delta G$__ 

Как и в предыдущем пункте всегда $С1 = С2$. Располагаемый доход частного сектора падает только во 2м периоде, а вот потребление будет падать в обоих периодах. Это значит что $CA1>0, CA2<0$.


__Пусть $\beta =1/(1+r)$  и изначально ожидалось, что $Y1-G1=Y2-G2$, но затем происходит рост госрасходов и в 1-м, и во 2-м периоде на $\Delta G$__ 


Как и в предыдущем пункте С1=С2 всегда. $Y1-G1=Y2-G2 = С1= С2$  по-прежнему удовлетворяет условиям равновесия (Эйлер и бюджетное ограничение). Так что $Ci$ упадет ровно на $\Delta G$, а $CAi=0$  останутся.



__Если государство ставит целью максимизацию общественного благосостояния:__

Гос-во знает, что потребитель выбирает $Ci$, удовлетворяя уравнению Эйлера и бюджетному ограничению. 

Получаются оптимальные решения $С1^*(G1,G2),  С2^*(G1,G2)$.


Задача максимизации для государства: 

$$ \max u(C1^*(G1,G2)))+ \beta u(C2^*(G1,G2)) + v(G1)+ \beta v(G2) $$

БО  государства (оно не может использовать больше товаров, чем есть в экономике)  

$$ Y2-G2+(1+r)(Y1-G1) > 0 $$
 
Если БО превратится в равенство, то это будет означать что для потребителей останется ровно одна доступная точка $С1=С2=0$. Вряд ли это когда-либо окажется оптимальным решением. Обычно в задачах, где есть потребление, мы по умолчанию предполагаем, что нулевое потребление невозможно, агент «умрет с голоду». В этой задаче это предположение неважно, но в целом оно часто используется.
