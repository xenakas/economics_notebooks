### Методология VAR.

####    1.  Комиссия Коулса: системы одновременных уравнений

Cпецификация модели производится с априорным разделением переменных на экзогенные и эндогенные.

Идентификация  достигается как результат накладывания большого количества ограничений.

При выявлении тех или иных отклонений от стандартных предположений метода наименьших квадратов модифицируется не модель, а метод оценивания.  

_Критика Лукаса и Симса_

Согласно Лукасу, традиционные структурные макромодели бесполезны
для целей симуляции политики, поскольку такие модели не принимают в расчет в явной форме ожидания экономических агентов.

Согласно Симсу, в моделях комиссии идентификация достигается за счет
произвольного объявления некоторых переменных экзогенными. 4

Однако в мире
агентов, поведение которых зависит от решения некоторых впередсмотрящих
межвременных оптимизационных моделей, никакая из переменных не может
считаться экзогенной. 
Ошибочная трактовка переменной как экзогенной
ведет к потере обратной связи и к возникновению кажущейся причинной связи.

Кроме того, в теоретическую априорную модель может быть включено
недостаточное количество переменных (и тогда возникает эффект пропущенных переменных) и недостаточное количество запаздываний.
 

####   2. Методология VAR

В связи с критикой Лукаса ключевой чертой новой классической макроэкономики стало разделение изменений в монетарной политике на два типа:
- изменения, которые агенты экономики предвидят правильно;
- изменения, которые являются неожиданными для агентов экономики.
Изменения первого типа должны производить нейтральные эффекты: про­
порциональные изменения цен и других номинальных переменных и отсутст­
вие влияния на реальные переменные. Неожиданные изменения (второго типа),
напротив, могут отражаться и на реальных переменных.

В методологии Симса акцент делается на исследование откликов системы
экономических показателей на неожиданные (шоковые) воздействия, которым подвергаются отдельные переменные.



В основе методологии VAR (как и методологии LSE) лежит общая структура вида:
$$ A y_t = C (L)y_{t-1} + \zeta_t $$
где $\zeta_t $ — инновационная последовательность i.i.d. случайных 
$ (k \times 1) $ векторов с нулевым математическим ожиданием.

Приведенная форма этой структуры имеет вид:
    $ (I- П (L)) y_t = u_t$

При этом $u_t$ также является инновационной последовательностью i.i.d.
случайных векторов, имеющих нулевое математическое ожидание и ковариационную матрицу $\Sigma$.


Вместо условия $u_t   \sim i.i.d.$ , конечно, достаточно, чтобы последовательность $u_t$  была векторным белым шумом с указанными характеристиками, что
можно обозначить как $u_t   \sim WN (0, Σ)$ .



_Векторное MA-представление   ряда  $y_t$_ 

Если выполнено условие стабильности, то

$$y_t = (I - П(L))^{-1} u_t  = u_t + \Psi_1 u_{t-1} + ... = \Psi(L)u_t$$

Тем самым, мы имеем возможность построения функций откликов обеих переменных на шоки инноваций                                 .





####    3. Фундаментальные инновации

Для преодоления затруднения при построении функций откликов  предполагают, что система изменяется благодаря воздействию некоррелированных между собой  “фундаментальных” инноваций             

Обычно предполагается, что все они имеют единичные дисперсии,  так что         $e_{1t}, ... , e_{kt}$                       – i.i.d.  с нулевым математическим ожиданием и единичной ковариационной матрицей  $I_k$ .

При этом предполагается, что инновации       $u_{1t}, ..., u_{kt}$                      являются линейными комбинациями ортогонализованных (фундаментальных) инноваций, так что: $u_t = D e_t$

                              



####    4. Функция импульсного отклика   (impulse response function – IRF ) 

Вычисляя последовательно,  на сколько изменяются значения $y_{1,t+s}, y_{2,t+s}...$           при изменении инноваций  $u_{1t}, u_{2t}$   на одно стандартное отклонение, для значений $s = 0, 1, …  $  получаем  функции откликов  на шоки инноваций. 

Из-за того, что общем случае          $Cov(u_{1t}, u_{2t})\neq 0$                       , возникают затруднения с интерпретацией  этих функций. 


Из-за перекрестной коррелированности инноваций в приведенной форме, невозможно полностью изолировать шок для  $u_{1t}$  от шока для  $u_{2t}$  



С экономической точки зрения, первоочередной интерес представляют 
реакции значений            $y_{1,t+s},..., y_{k,t+s}$                                   на единичные импульсные изменения отдельных фундаментальных инноваций          $e_{1t}, ...,e_{kt}$           
 при фиксированных значениях всех остальных фундаментальных инноваций 
во все моменты времени.  





У стабильной  VAR функции импульсного отклика $IRF(h)$ при $ h \to \infty$  стремятся
к нулю.




####    5. Проблема идентификации структурной формы VAR на основе приведенной формы

Если матрица $\Sigma$ известна, для нахождения четырех  неизвестных  $d_{11} , d_{12} , d_{21} , d_{22}$   имеется всего три уравнения.   Матрицу  $D$  идентифицировать невозможно, 
если не накладывать априорных ограничений на ее структуру.



Предложение Симса: упорядочивание инноваций в системе.

Получаем структурную модель:

$$A y_t = A  П  y_{t-1} + A P^* B e_t  = C y_{t-1} + B e_t$$

 Сначала определяется значение $y_{1t}$  из первого уравнения, 
а затем на его основе определяется значение  $y_{2t}$  из второго уравнения.  (“Переменная  $y_{1t}$  входит в уравнения первой” )


$$\Sigma = Cov(u_t) = Cov(A^{-1} B e_t) = A^{-1} B Cov(e_t) (A^{-1}B)^{T} = A^{-1} B  B^{T} (A^{-1})^{T} $$  


Оценив приведенную VAR, мы можем получить и оценку $\hat{\Sigma}$         ковариационной матрицы    $\Sigma$.

Матрица     $\hat{\Sigma}$    симметрична, и поэтому достаточно оценить 
$\frac{k^2 +k}{2} $
ее элементов.

Общее количество неизвестных элементов в матрицах    $\hat{A}$   и         $\hat{B}$ равно
        $2k^2$.
        
Поэтому идентификация возможна лишь  при наложении на матрицы   $A$    и    $B$    достаточного количества ограничений. 


Методология Симса, которую мы применили к VAR(2),  фактически основана на следующем результате (разложение Холецкого):
$$\Sigma = P P^T$$
где   $P$   –  нижняя треугольная матрица с положительными диагональными элементами,  причем такая матрица       единственна.


В роли матрицы     $\Sigma$   у нас будет выступать матрица $D$ из соотношения     $u_t = D e_t$, так что $$D = P P^T, A= (P^*)^{-1},  B = I (p_{11} ... p_{kk}) $$



Если наложить на матрицы $A$ и $B$  следующие ограничения:
- $A$  – нижняя треугольная матрица с единицами на диагонали;
- $B$  – диагональная матрица с положительными диагональными элементами,
то уравнение 
$$\Sigma = A^{-1} B  B^{T} (A^{-1})^{T}$$
имеет единственное решение, т.е. имеет место точная идентифицируемость  матриц $A$ и $B$.

Диагональные элементы $b_{jj}$   матрицы $B$ можно рассматривать как с.к.о. инновации в $i$-м уравнении структуры, а форма матрицы $A$ соответствует рекурсивной системе .


####    6. Рекурсивная система

В первое уравнение с текущим значением входит только одна переменная   $y_{1t}$.

Во второе уравнение с текущими значениями входят обе переменные   $y_{1t}$  и  $y_{2t}$.  

...

В последнее, $k$-е уравнение с текущими значениями входят все переменные   $y_{1t} , y_{2t} , . . . , y_{kt}$


Такой порядок вхождения переменных интерпретируется как последовательное включение переменных  в порядке возрастания их эндогенности,  так что последней в систему включается наиболее эндогенная переменная. 
 
 


Можно  оценивать полученную рекурсивную систему обычным методом наименьших квадратов.

####    9. Декомпозиция дисперсий ошибок прогнозов

Ошибка прогноза по VAR на один шаг вперед:
$$e_{t+1|t} = y_{t+1} - \hat{y}_{t+1|t} = u_{t+1} = D e_{t+1}$$
имеет среднеквадратическую ошибку $E[e_{t+1|t} e_{t+1|t}^T = DD^T]$

Получим декомпозиции дисперсий ошибок прогноза 
                – разложение каждой из них на две компоненты: 

Дисперсия ошибки прогноза на один шаг вперед ряда $y_1$:
$D(e_{1,t+1|t}) = d^2_{11} + d^2_{12}$

Дисперсия ошибки прогноза на один шаг вперед ряда $y_2$:
$D(e_{2,t+1|t}) = d^2_{21} + d^2_{22}$


$d_{11} e_{1,t+1}$ – составляющая ошибки прогноза на один шаг вперед ряда $y_1$, 
обусловленная фундаментальной инновацией  $e_{1,t+1}$


$d_{11}^2$ – составляющая дисперсии ошибки прогноза на один шаг вперед
 ряда $y_1$, обусловленная фундаментальной инновацией  $e_{1,t+1}$ 


$\frac{d^2_{11}}{d^2_{11} + d^2_{12}}$ – доля дисперсии ошибки прогноза на один шаг вперед 
 ряда $y_1$, обусловленная фундаментальной инновацией $e_{1,t+1}$
 
 Дисперсия ошибки прогноза значения  $y_{s,t+h}$ , сделанного в момент $t$ , представляется  в виде суммы неотрицательных слагаемых, каждое из которых определяется исключительно  шоком соответствующей фундаментальной инновации в момент  $t$.
 
Обычно результат такой декомпозиции представляется как перечень долей каждого из слагаемых в общей сумме. 


####    10. Проблема, возникающая при построении импульсных откликов:

В методологии VAR к имеющимся данным подбирается модель 
     авторегрессии конечного порядка, и оцененная модель
     обращается для получения MA-коэффициентов, составляющих 
     импульсную функцию откликов.

Использование VAR конечного порядка основано на предположении о том, что AR-представление существует   и может быть приближено VAR моделью конечного (быть может, достаточно высокого) порядка.
 
Соответственно, если процесс необратим и не имеет VAR-представления, то в такой ситуации представляется бессмысленным использовать VAR методологию для построения импульсных откликов.



Для практически любого стационарного векторного процесса:
подобранная по достаточно длинному периоду наблюдений модель VAR имеет VMA представление, которое асимптотически является корректным,
__хотя само VAR представление процесса может и не существовать__!


При наличии единичных корней или корней, близких к единице
 импульсные отклики, полученные на основании оцененной приведенной VAR без ограничений, в долговременном плане сходятся в пределе не к истинным откликам, а к некоторым случайным величинам.
 
Предельное распределение не является нормальным и несимметрично.

Предсказания на основании нестационарной VAR без ограничений не сходятся к оптимальным прогнозам  (по крайней мере, на больших горизонтах прогнозов), а  дисперсия оптимального прогноза, как правило занижается.

####    11. Построение функции импульсных откликов в случае некоинтегрированной VAR

Если ряды  имеют порядок I(1) и не коинтегрированы

В этом случае   ряд  разностей     $\Delta y_t$         представляется в виде  
$$\Delta y_t = (1-L) y_t = C(L) e_t $$
где $C(L)$ - – функция импульсных откликов VAR в  разностях

Функцию импульсных откликов в уровнях  (accumulated responses), так что 
 $ y_t = C^*(L) e_t$  можно восстановить последовательным накоплением импульсных откликов в разностях

####    12. Построение  функции импульсных откликов в случае коинтегрированной VAR


Если имеет место коинтегрированность рядов, образующих VAR, и оценивается модель VAR, учитывающая ограничения, накладываемые коинтеграцией, то тогда  построенные по оцененной модели импульсные отклики и декомпозиции дисперсий прогнозов состоятельны, а прогнозы асимптотически оптимальны. 


Правда, для этого надо или 
правильно специфицировать ранг коинтеграции  или
получить для него состоятельную оценку. 


Использование при оценивании ECM различных предположений о ранге коинтеграции  приводит к различным результатам в отношении функций импульсного отклика.

Это весьма затрудняет интерпретацию  результатов VAR анализа при обращении к данным экономической статистики.



####    13. Построение функций импульсных откликов в случае, когда статистические данные порождены моделью векторного скользящего среднего

_Долговременные ограничения_

Поскольку экономическая теория часто не в состоянии предложить разумные одновременные ограничения (а чем больше переменных включается в систему, тем большее количество таких ограничений требуется для идентифицируемости), в работах  Blanchard and Quah (1989), Shapiro and Watson (1988),  Galí (1992) было предложено накладывать ограничения на долговременные свойства системы. 

 Соответствующие долговременные ограничения (long run) обычно базируются на некоторых постулатах нейтральности. 


Подход, о котором мы будем теперь говорить, основан на структурной модели скользящего среднего (SMA), получаемой на основе модели SVAR
(если  VAR стационарна)

Долговременные ограничения  предполагают отсутствие не    немедленной, а накопленной реакции. Соответственно, в случае двух рядов “нейтральность”  влияния  шока   $e_{2t}$         на      $y_{1t}$   понимается как выполнение соотношения:
$\theta_{12}(1) = \sum \theta_{12, j} =0  $


Это приводит к  нелинейному ограничению на коэффициенты SVAR.
