## 4. Модель AA-DD

Фокус на внешнем секторе и обменном курсе:

- Инвестиции равны константе (не как IS-LM)
- Добавлены ожидания будущего – они задействованы в условии отсутствия арбитража для финансовых рынков
- Цены меняются только в LR, выполняется ППС
- Вместо моментов времени $t$ и $t+1$ (как раньше) моменты старта, SR равновесия и LR равновесия (LR реализуется как бы в  $t+1$)
- Предполагаем мгновенный переход от старта к SR
- Переход между SR и LR не моделируется

### AA

Денежный рынок: 

- Процентная ставка = результат равновесия спроса и предложения
денег 

- Классическая теория денег:

$$m^s = M^s/P \\  M^d = L(Y,i) $$

Для упрощения предполагаем единую на все инструменты ставку и единый денежный агрегат (он же денежная масса)

Валютный рынок:

- Определяется отсутствием арбитража между финансовыми вложениями в домашние и зарубежные активы 
- Все вложения должны давать одинаковый приведенный доход
- Домашняя cтавка минус потери на девальвации = зарубежная ставка

$$( 1+ i ) = (1 + i^* ) \frac{ S^e}{S} $$
 

Условие отсутствия арбитража называется «непокрытый паритет процентных ставок»

AA:

Комбинируем денежный и валютный рынки: 
- Склейка по процентной ставке
- Ставка определятся на денежном рынке, а затем проецируется через условие отсутствия арбитража на валютный рынок, давая равновесный спотовый обменный курс

Какая связь с реальным сектором: 

При увеличении выпуска Y у агентов растет спрос на деньги (больше транзакций, больше денег нужно при тех же ценах), ставка на денежном рынке растет (все занимают еще денег), наши активы более доходными, спрос на нашу валюту растет, растет цена нашей валюты, а обменный курс S снижается

АА – это кривая равновесия на 
рынках активов, валютном и
денежном

_Интуиция - одна и та же денежная масса будет менее востребована в меньшей по размеру экономике (просто не нужно так много денег при тех же ценах), эта невостребованность проецируется на валютный рынок через низкие ставки. Итог: нац. валюта дешевле_

Cдвиг кривой АА вправо/вверх:

- повышение предложения денег (больше денег – дешевле валюта при неизменном размере экономики Y)
- рост зарубежных проц. ставок 
- рост ожидаемой девальвации в будущем $S^e$
- снижение уровня цен (так как при этом растет реальное предложение денег)
- развитие финансовых рынков (так как транзакции проходят быстрее, а деньги заменяются другими инструментами)

### DD

AD:

$$ AD = C(Y) + I + G + CA(SP^*/P, Y^*) $$


Совокупный спрос определяет, сколько наших товаров и услуг (Y) будет куплено при каждом уровне обменного курса (S) 

__Условия Маршалла-Лернера__

$$ CA (RER, Y, Y^*) = EX ( RER, Y^*) – IM (RER, Y) $$

Условия МЛ: $\partial CA/\partial RER > 0$

Баланс внешнего сектора улучшается при реальном обесценении валюты



Интуиция: чем дешевле валюта, тем дешевле наши товары по сравнению с зарубежными (тем выше относительный спрос на наши товары, тем больше экспорт и меньше импорт)


Почему может упасть спрос:
- Мировой финансовый кризис 
- лопнувшие пузыри делают население беднее $C \downarrow$
- ограничение кредита фирмам $I \downarrow$
- падение цен на нефть ведет $EX \downarrow$


DD: 


DD – это кривая равновесия на рынке товаров

Интуиция: дешевая домашняя валюта  улучшает торговый баланс, а, следовательно, и ВВП

$$ Y = C(Y_d) + I + G + EX(SP^*/P) – IM(SP^*/P, Y) $$ 


- Потребление  $C(Y_d ) = C_0 +c*(Y-T)$
- Экспорт положительно зависит от q, мпорт отрицательно зависит от q и положительно от Y
- Чистый экспорт $NX(SP*/P, Y) = NX_0 + a \ SP^*/P - mY$ 




Cдвиг кривой DD вправо/вниз: 

- Повышение спроса AD: положительный фактор должен увеличить выпуск при неизменном обменном курсе
- Рост госрасходов, трансферт
- Снижение налогов и ожидаемое снижение налогов в будущем 
- Рост инвестиций, ожидания роста богатства, рост отдачи на инвестиции, улучшение инвестиционного климата 
- Рост цен за границей, рост внешнего спроса на наши товары и услуги 
- Снижение уровня цен, так как удешевляет наши товары по сравнению с иностранными

### AA-DD



Комбинируем AA равновесие на
денежном и валютном рынках и DD – равновесие на рынке
товаров. Курс S (плавающий) задается этим пересечением

Если пересечение не на $Y^n$, то с
течением времени AA и DD
медленно подстраиваются за счет
изменения цен, но эта подстройка требует
длительного времени 


> Adjustment to a new equilibrium follows the principle that adjustment in the asset markets occurs much more rapidly than adjustment in the goods and services market. Thus adjustment to the AA curve will always occur before adjustment to the DD curve.

Уменьшение совокупного спроса сдвигает кривую DD влево/вверх
- в краткосрочном периоде будет рецессия и обесценение валюты 
- в долгосрочном периоде цены должны  подстроиться, чтобы сохранить выпуск постоянным 



### Государственная политика во время кризиса

Модели делят на фискальную и монетарную (и та, и другая могут стимулировать экономику, сглаживая кризис)

Фискальная – рост госрасходов (и снижение налогов) 
- В реальности есть еще социалка и автоматические стабилизаторы (напр. пособия по безработице)
- Медленная, т.к. нужно принять закон о бюджете
- Политические интересы могут тормозить принятие законов и искажать структуру

Монетарная – дешевые деньги
- В основном за счет снижения ставок ЦБ
- Быстрая, обычно не требует решений правительства и парламента
- Ограничена для стран с высокой инфляцией и/или нестабильным валютным рынком

Структурные реформы:
- Кризис это плохое время для реформ (нет денег)
- Но в другие периоды политики не хотят их проводить

Растет предложение денег, и кривая АА сдвигается вправо/вверх: 

Если временно:
- Долгосрочное равновесие не меняется
- Значит ожидания будущего не меняются
- SR равновесие в точке 2

Если перманентно:
- Долгосрочное равновесие меняется 
- Ожидания будущего меняются
- В данном случае больше рублей в LR, значит рубль сравнительно дешевле по сравнению с временным ростом $M^s$, pначит $E^e$ растет
- Значит АА сдвигается еще дальше


Если шок перманентный и LR-равновесие изменилось, то решаем модель с конца:

- Понимаем, какое примерно равновесие будет в LR, учитывая $Y = Y^{FE}$
- ППС
- Понимаем, какой примерно $S_{LR}$ ожидается
- Соответственно корректируем ожидания агентов $S^e$ уже в SR ( поэтому эффект может быть очень быстрый и вполне реальный, даже если в экономике ничего не поменялось, кроме самих ожиданий)
- Зная $S^e$, находим SR-равновесие


Может изменяться и долгосрочное равновесие $Y_{FE}$:
- Технологический прогресс
- Эффективность рынков труда капитала
- Эффективность рынка продукции и международной торговли
- Демография и образование
- Цены на сырье и другие издержки
- Сегодняшняя производительность (не путать с будущей производительностью, которая влияет на совокупный спрос через инвестиции)


Частный случай – ловушка ликвидности:

Ловушка ликвидности возникает, когда
номинальные ставки снижаются до нуля, а
ЦБ не может поощрять людей удерживать
больше ликвидных активов (денег).
 

- Если процентные ставки близки к нулю, то  ожидания относительно обменного курса (и инфляции) постоянные и фиксированные иностранные ставки, спот-курс эффективно фиксируется
- Покупка ЦБ внутренних активов ЦБ не снижает ставку и не обесценивает курс
- Денежная политика становится неэффективной



### Политика ЦБ

- Операции с резервами изменяют денежную базу 
- Денежная база определяется как сумма национальной валюты в обращении 

Инструменты ЦБ для изменения денежной базы:

- Валютные интервенции: ЦБ покупает или продает иностранную валюту в обмен на внутреннюю валюту 
- Операции на открытом рынке: покупка или продажа государственных и других ценных бумаг на открытом рынке.
- Дисконтные кредиты коммерческим банкам: кредиты банкам в установленном порядке с известными процентными ставками

Интервенции ЦБ на валютном рынке:

Когда ЦБ покупает иностранную валюту на рынке
- Увеличение официальных резервов
- Деньги в обращении растут, так как ЦБ расплачивается свеже напечатанными рублями
- Увеличивается денежная масса
- Смягчение денежной политики

Когда ЦБ продает иностранную валюту
- Уменьшение денежного предложения, потому что национальная валюта выплачивается в ЦБ экономическими агентами в обмен на валюту, следовательно, она удаляется из экономики
- Ужесточение денежной политики

Стерилизация - операция на других рынках, направленная на
компенсацию эффектов на денежную массу от
покупки/продажи валюты


### Макрополитика в модели AA-DD при фиксированном курсе


Фиксированный курс  $S_0$: 





Зависимости на финансовых рынках разворачиваются: курс определяет ставку, а та, в свою очередь, определяет денежную массу



Следствие UIP: $i = i^*$, поскольку нет валютного риска

Денежная масса будет эндогенной! То есть будет определяться внутри модели 



Примечание к организации режима фикс. курса:

Пока ЦБ выполняет обещание  в любой
момент продавать и покупать валюту по $S_0$, рыночный курс
автоматически устанавливается на $S_0$
- Если вдруг рыночный $S>S_0$, то все участники рынка покупают валюту не на рынке, а у ЦБ (потому что дешевле!)
- В результате ЦБ тратит свои валютные резервы, а рынок получает приток иностранной валюты (из резервов)
- И цена на валюту снижается, то есть $S$ приближается к $S_0$, пока не сравняется




Если денежной массы больше или
меньше, чем надо для равновесия

- рыночная ставка захочет поменяться, а инвесторы соответственно захотят перебалансировать портфели, что вызовет изменение спроса на валюты и тенденцию к изменению рыночных цен валют, то есть курса
- но пока есть ЦБ-торговля по $S_0$, курс не может измениться. Вместо этого меняется денежная масса, когда инвесторы торгуют валютой с ЦБ


- кривая денежного рынка (AA) теряет самостоятельность

Если пересечение АА и DD смещается в точку, где курс S отличается от первоначального, то
- ЦБ торгует валютой
- Изменяется денежная масса
- Кривая АА перемещается, пока не восстановится фиксированный курс

- Как итог полностью неэффективная монетарная политика, и очень эффективная фискальная


Девальвация как политика при фиксированном курсе: 
- для девальвации ЦБ устанавливает целевой курс $S_1$ выше первоначального $S_0$
- агенты продают $FX$ в ЦБ (так как он предлагает самую высокую цену на рынке), пока рыночная цена FX не сравняется с курсом ЦБ. 
- увеличивается денежная база и снижаются процентные ставки
- официальные международные резервные активы в ЦБ растут 
- домашние товары становятся дешевле на мировом рынке, поэтому растет совокупный спрос и производство

НО! девальвация работает на полную, только если она происходит неожиданно
- в случае постоянно идущей девальвации ожидания опережают, и сдвигают кривую АА, не давая вырасти денежной базе 
- может быть даже обратный эффект - рынок начинает скупать FX из-за страха будущих девальваций 

- Поскольку официальный $S_0$ еще не изменен (девальвация не объявлена), ЦБ должен защищать фиксированный курс. Он делает это, продавая резервы для удовлетворения выросшего рыночного спроса на FX.
- Денежная масса уменьшается, сдвигая назад AA и восстанавливая равновесие на уровне привязки
- Но при этом резервы могут закончиться, и ЦБ не сможет защитить привязку, увеличение процентной ставки напрягает банковскую систему



Если ЦБ попытается повысить M^s:

- Рыночная ставка начинает снижаться
- Домашние активы теряют доходность по сравнению с иностранными, инвесторы перекладываются в иностранные активы, растет спрос на иностранную валюту
- Рыночный курс $S$ начинает повышаться, и становится $S>S_0$
- Но всегда есть $S_0$ от ЦБ - поэтому все покупают иностранную валюту у ЦБ, а не на рынке
- Снижаются резервы ЦБ (дисциплина!) и одновременно снижается предложение денег в экономике, так как ЦБ продает валюту и получает рубли (которые автоматически выходят из обращения) 
- Покупки валюты у ЦБ будут продолжаться, пока ставки опять не сравняются, то есть пока все не вернется к прежнему Ms
- Корректировка происходит быстро, и фактический сдвиг кривой не  происходит 

__В режиме фиксированного курса денежно-кредитная политика полностью неэффективна!__




Фискальная политика при фикс. курсе:
- G увеличивается, DD сдвигается в DD’
- S хочет укрепиться, ЦБ предотвращает укрепление и покупает с рынка иностранную валюту 
- Ms увеличивается, AA сдвигается в AA’

Не имеет значения, является ли
увеличение G временным или
перманентным

__В режиме фиксированного курса
фискальная политика очень
эффективна!__



В долгосрочной перспективе невозможно поддерживать
увеличение Ms:
- Любое увеличение Ms тут же автоматически откатывается назад и одновременно уменьшает резервы ЦБ
- Когда-нибудь резервы закончатся, и вместе с ними закончится возможность поддерживать режим фиксированного курса

##### Задача

_Монетарная политика более эффективна в отношении стимулирования выпуска при режиме плавающего обменного курса, чем при режиме фиксированного обменного курса_.

- Монетарная политика = изменение предложения денег = сдвиг кривой АА
- Стимулирование выпуска = увеличение уровня ВВП. Отсюда следует, что рассматриваем увеличение предложения денег, т.е. сдвиг кривой АА вправо-вверх.
- В долгосрочной перспективе макрополитики не могут стимулировать выпуск, который устанавливается на своем долгосрочном равновесном уровне независимо от политики.
- Режим фикс.курса = полная неэффективность монетарной политики, рост Ms полностью компенсируется выкупом рынком резервов у ЦБ, и соответствующим падением Ms обратно в краткосрочной перспективе. То есть стимулирования нет
- Режим плав.курса = кривая АА успешно сдвигается вправо-вверх, то есть точка ее пересечения с DD переезжает вправо-вверх по кривой DD. Соответственно, выпуск растет, и стимулирование есть.
